- Í∑∏ Ïô∏ Ï∂úÎ†•Î¨∏ Î∞©ÏãùÏùÄ depositÍ≥º ÎèôÏùºÌïú Î∞©ÏãùÏúºÎ°ú ÏûëÏÑ±Ìï©ÎãàÎã§.

```python
import random # Í≥ÑÏ¢åÎ≤àÌò∏ ÎûúÎç§ ÏÉùÏÑ±Ïö©

class Account:
    # Í≥ÑÏ¢å Í≥ÑÏàò Ï∂úÎ†•ÏúÑÌïú Î≥ÄÏàò
    account_count = 0 # Î≥ÄÏàò Ï¥àÍ∏∞Ìôî

    # ÏÉùÏÑ±Ïûê ÏÉùÏÑ±
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # Í≥ÑÏ¢å ÏµúÏ¥à Í∞úÏÑ§Ïãú 0ÏõêÏù¥ÎØÄÎ°ú
        self.bank_name = 'SCÏùÄÌñâ' # ÏùÄÌñâ Ïù¥Î¶Ñ Í≥†Ï†ï
        self.account_number = self._generate_account_number()

        # ÏÉà Í≥ÑÏ¢å ÏÉùÏÑ±Ïãú ÎßàÎã§ Í≥ÑÏ¢å Í∞úÏàò +1
        Account.account_count += 1


    # ÎûúÎç§ Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Î©îÏÑúÎìú
    def _generate_account_number(self):
        """
        Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Í∑úÏπô : 3ÏûêÎ¶¨-2ÏûêÎ¶¨-6ÏûêÎ¶¨
        randomÌï®ÏàòÏùò randint Î©îÏÑúÎìú ÌôúÏö©
        Ïà´ÏûêÏùò Î≤îÏúÑÎäî ÏûêÎ¶¨ÏàòÏóê Îî∞ÎùºÏÑ§Ï†ï
        Íµ¨Ï°∞Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥¥Í∏∞ ÏúÑÌï¥ fstringÏÇ¨Ïö© (ÏùºÎ∞ò printÎ¨∏Ïù¥Î©¥ ""ÏôÄ ,Î•º ÏÑûÏñ¥Ïïº Ìï¥ÏÑú Í∞ÄÎèÖÏÑ± Îñ®Ïñ¥Ïßê)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # Í≥ÑÏ¢å Í∞úÏàò Ï∂úÎ†• Î©îÏÑúÎìú Ï∂îÍ∞Ä
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. ÏûÖÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def deposit(self, amount):
        # ÏûÖÍ∏àÏ†Ñ ÌòÑÏû¨ ÏûîÏï° Ï†ïÎ≥¥ Ï∂úÎ†•
        print("ÏûÖÍ∏à Ï†Ñ ÏûîÏï°:", "{:,}Ïõê".format(self.balance)) # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)

        # ÏµúÏÜå ÏûÖÍ∏àÏï° Í∏∞Ï§Ä Ï∂©Ï°± Ïó¨Î∂Ä ÌåêÎã®
        if amount >= 1:
            self.balance += amount # ÏûîÍ≥†Ïóê ÏûÖÍ∏àÏï° Ï∂îÍ∞Ä
            # Í≤∞Í≥º Ï∂úÎ†•
            print(f"ÏûÖÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,}Ïõê ÏûÖÎãàÎã§.") # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)
        # ÏµúÏÜå ÏûÖÍ∏àÏï° ÎØ∏Ï∂©Ï°±Ïãú Î¨∏Íµ¨ Ï∂úÎ†•
        else:
            print(f"ÏûÖÍ∏à Ïã§Ìå®! - ÏûÖÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏµúÏÜå ÏûÖÍ∏àÏï°ÏùÄ 1ÏõêÏûÖÎãàÎã§. Ïû¨ÏûÖÍ∏àÏùÑ ÏãúÎèÑÌï¥Ï£ºÏÑ∏Ïöî!")


    # Q5. Ï∂úÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def withdraw(self, amount):
        # ÌòÑÏû¨ ÏûîÍ≥† Ï∂úÎ†•
        print("ÌòÑÏû¨ ÏûîÍ≥† :", "{:,}Ïõê".format(self.balance))

        # Ï∂úÍ≥† ÏöîÏ≤≠ Í∏àÏï°Ïóê ÎåÄÌïú Ï∂úÍ∏à Ïó¨Î∂Ä Í≤ÄÏ¶ù
        if amount <= self.balance:
            self.balance -= amount # Ï∂úÍ∏àÏöîÏ≤≠Í∏àÏï°ÏùÑ ÏûîÍ≥†Ïóê Î∞òÏòÅ(Ï∞®Í∞ê)
            print(f"Ï∂úÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,}Ïõê ÏûÖÎãàÎã§.")
        else:
            print(f"Ï∂úÍ∏à Ïã§Ìå®! - Ï∂úÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏûîÏï°Ïù¥ Î∂ÄÏ°±Ìï©ÎãàÎã§.")




# ÌÖåÏä§Ìä∏
customer1 = Account("ÍπÄÏäπÏàú") # Ïã†Í∑úÍ≥†Í∞ù ÏÉùÏÑ±
customer2 = Account("ÍπÄÏïÑÏù¥Ìé†") # Ïã†Í∑úÍ≥†Í∞ù ÏÉùÏÑ±

# Í≥†Í∞ù 1 20,000Ïõê ÏûÖÍ∏à
customer1.deposit(20000)

# Í≥†Í∞ù 1 10,000Ïõê Ï∂úÍ∏à ÏãúÎèÑ - ÏÑ±Í≥µ Î©îÏãúÏßÄ Ï∂úÎ†•
customer1.withdraw(10000)

# Í≥†Í∞ù 2 50000Ïõê ÏûÖÍ∏à
customer2.deposit(50000)

# Í≥†Í∞ù 2 60000Ïõê Ï∂úÍ∏à ÏãúÎèÑ - Ïã§Ìå® Î©îÏãúÏßÄ Ï∂úÎ†•
customer2.withdraw(60000)
```

```python
import random # Í≥ÑÏ¢åÎ≤àÌò∏ ÎûúÎç§ ÏÉùÏÑ±Ïö©

class Account:
    # Í≥ÑÏ¢å Í≥ÑÏàò Ï∂úÎ†•ÏúÑÌïú Î≥ÄÏàò
    account_count = 0 # Î≥ÄÏàò Ï¥àÍ∏∞Ìôî

    # ÏÉùÏÑ±Ïûê ÏÉùÏÑ±
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # Í≥ÑÏ¢å ÏµúÏ¥à Í∞úÏÑ§Ïãú 0ÏõêÏù¥ÎØÄÎ°ú
        self.bank_name = 'SCÏùÄÌñâ' # ÏùÄÌñâ Ïù¥Î¶Ñ Í≥†Ï†ï
        self.account_number = self._generate_account_number()

        # ÏÉà Í≥ÑÏ¢å ÏÉùÏÑ±Ïãú ÎßàÎã§ Í≥ÑÏ¢å Í∞úÏàò +1
        Account.account_count += 1


    # ÎûúÎç§ Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Î©îÏÑúÎìú
    def _generate_account_number(self):
        """
        Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Í∑úÏπô : 3ÏûêÎ¶¨-2ÏûêÎ¶¨-6ÏûêÎ¶¨
        randomÌï®ÏàòÏùò randint Î©îÏÑúÎìú ÌôúÏö©
        Ïà´ÏûêÏùò Î≤îÏúÑÎäî ÏûêÎ¶¨ÏàòÏóê Îî∞ÎùºÏÑ§Ï†ï
        Íµ¨Ï°∞Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥¥Í∏∞ ÏúÑÌï¥ fstringÏÇ¨Ïö© (ÏùºÎ∞ò printÎ¨∏Ïù¥Î©¥ ""ÏôÄ ,Î•º ÏÑûÏñ¥Ïïº Ìï¥ÏÑú Í∞ÄÎèÖÏÑ± Îñ®Ïñ¥Ïßê)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # Í≥ÑÏ¢å Í∞úÏàò Ï∂úÎ†• Î©îÏÑúÎìú Ï∂îÍ∞Ä
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. ÏûÖÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def deposit(self, amount):
        # ÏûÖÍ∏àÏ†Ñ ÌòÑÏû¨ ÏûîÏï° Ï†ïÎ≥¥ Ï∂úÎ†•
        print("ÏûÖÍ∏à Ï†Ñ ÏûîÏï°:", "{:,}Ïõê".format(self.balance)) # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)

        # ÏµúÏÜå ÏûÖÍ∏àÏï° Í∏∞Ï§Ä Ï∂©Ï°± Ïó¨Î∂Ä ÌåêÎã®
        if amount >= 1:
            self.balance += amount # ÏûîÍ≥†Ïóê ÏûÖÍ∏àÏï° Ï∂îÍ∞Ä
            # Í≤∞Í≥º Ï∂úÎ†•
            print(f"ÏûÖÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,}Ïõê ÏûÖÎãàÎã§.") # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)
        # ÏµúÏÜå ÏûÖÍ∏àÏï° ÎØ∏Ï∂©Ï°±Ïãú Î¨∏Íµ¨ Ï∂úÎ†•
        else:
            print(f"ÏûÖÍ∏à Ïã§Ìå®! - ÏûÖÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏµúÏÜå ÏûÖÍ∏àÏï°ÏùÄ 1ÏõêÏûÖÎãàÎã§. Ïû¨ÏûÖÍ∏àÏùÑ ÏãúÎèÑÌï¥Ï£ºÏÑ∏Ïöî!")


    # Q5. Ï∂úÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def withdraw(self, amount):
        # ÌòÑÏû¨ ÏûîÍ≥† Ï∂úÎ†•
        print("ÌòÑÏû¨ ÏûîÍ≥† :", "{:,}Ïõê".format(self.balance))

        # Ï∂úÍ≥† ÏöîÏ≤≠ Í∏àÏï°Ïóê ÎåÄÌïú Ï∂úÍ∏à Ïó¨Î∂Ä Í≤ÄÏ¶ù
        if amount <= self.balance:
            self.balance -= amount # Ï∂úÍ∏àÏöîÏ≤≠Í∏àÏï°ÏùÑ ÏûîÍ≥†Ïóê Î∞òÏòÅ(Ï∞®Í∞ê)
            print(f"Ï∂úÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,}Ïõê ÏûÖÎãàÎã§.")
        else:
            print(f"Ï∂úÍ∏à Ïã§Ìå®! - Ï∂úÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏûîÏï°Ïù¥ Î∂ÄÏ°±Ìï©ÎãàÎã§.")


    # Q6. Ï†ïÎ≥¥Ï∂úÎ†• Î©îÏÑúÎìú Íµ¨ÌòÑÌïòÍ∏∞
    def display_info(self):
        # Í∞ÅÍ∞ÅÏùò Ï†ïÎ≥¥Î•º Ï∂úÎ†•ÌïòÎäî printÎ¨∏ ÏûëÏÑ±
        print(f"ÏùÄÌñâ Ïù¥Î¶Ñ: {self.bank_name}")
        print(f"ÏòàÍ∏àÏ£º: {self.account_name}")
        print(f"Í≥ÑÏ¢åÎ≤àÌò∏: {self.account_number}")
        print(f"ÏûîÏï°: {self.balance:,}Ïõê")



# ÌÖåÏä§Ìä∏
customer1 = Account("ÍπÄÏäπÏàú") # Ïã†Í∑úÍ≥†Í∞ù ÏÉùÏÑ±


# Í≥†Í∞ù 1 20,000Ïõê ÏûÖÍ∏à
customer1.deposit(20000)
customer1.display_info()
```

- Ïù¥Ï†Ñ Î¨∏Ï†úÏóêÏÑú Íµ¨ÌòÑÌïú ÏΩîÎìúÎì§Í≥º Ïù∏Ïä§ÌÑ¥Ïä§ Ï†ïÎ≥¥Îì§ÏùÑ Í∞ÄÏ†∏ÏôÄ Î∞îÎ°ú Ï∂úÎ†•ÌïòÍ≤å Ìï®ÏúºÎ°úÏç® display_info Î©îÏÑúÎìúÎ•º ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Íµ¨ÌòÑÌñàÏäµÎãàÎã§.

## Q7. Ïù¥Ïûê ÏßÄÍ∏âÌïòÍ∏∞ Í∏∞Îä• Íµ¨ÌòÑ : ÏûÖÍ∏à ÌöüÏàòÍ∞Ä 5ÌöåÍ∞Ä Îê†Îïå ÏûîÍ≥†Î•º Í∏∞Ï§ÄÏúºÎ°ú 1%Ïùò Ïù¥ÏûêÍ∞Ä ÏûîÍ≥†Ïóê Ï∂îÍ∞ÄÎêòÎèÑÎ°ù ÏΩîÎìúÎ•º Î≥ÄÍ≤ΩÌï¥Ïïº Ìï©ÎãàÎã§.

- Ï≤òÏùå ÌÅ¥ÎûòÏä§ Íµ¨Ï°∞Î•º Í∑∏Î¶¥Îïå Íµ¨ÏÉÅÌïú Î∞îÏôÄ Í∞ôÏù¥ deposit Î©îÏÑúÎìú ÏïÑÎûò Ïù¥ÏûêÎ•º ÏßÄÍ∏âÌïòÎäî Í∏∞Îä•ÏùÑ Íµ¨ÌòÑÌñàÏäµÎãàÎã§.
- Í∏∞Ï°¥ Î©îÏÑúÎìú ÌïòÏúÑÏóê Í∏∞Îä•ÏùÑ Íµ¨ÏÑ±Ìïú Ïù¥Ïú†Îäî Í∏∞Îä•Ïùò Íµ¨Î∂ÑÏùÑ Î™ÖÌôïÌïòÍ≤å ÌïòÍ∏∞ ÏúÑÌï® ÏûÖÎãàÎã§.
- ÏûÖÍ∏àÏù¥ÎûÄ Ïï°ÏÖòÏù¥ ÏùºÏñ¥ÎÇ†Îïå Î∞úÏÉùÌïòÎäî Ï∂îÍ∞ÄÏ†ÅÏù∏ Ïï°ÏÖòÏù¥Í∏∞ ÎïåÎ¨∏Ïóê depositÏù¥ Î∞úÏÉùÌïòÎäî ÌöüÏàòÏóê Îî∞Îùº ÏûëÎèôÌïòÎèÑÎ°ù Íµ¨ÌòÑÌñàÏäµÎãàÎã§.
- Ïù¥ Í≤ΩÏö∞ Îß§ ÏûÖÍ∏àÎßàÎã§ Ïù¥Ïûê ÏßÄÍ∏â Ï°∞Í±¥ÏùÑ Í≤ÄÌÜ†ÌïòÎØÄÎ°ú Î≥ÑÎèÑÎ°ú Í∏∞Îä•ÏùÑ Î∂àÎü¨ÏôÄÏÑú ÏÇ¨Ïö©Ìï† ÌïÑÏöî ÏóÜÏù¥ ÏûêÎèôÏúºÎ°ú ÏûëÎèôÌïòÎèÑÎ°ù Ìï† Ïàò ÏûàÏäµÎãàÎã§.

---

- Í∏∞Îä• Íµ¨ÌòÑÏùÑ ÏúÑÌï¥ ÌïÑÏöîÌïú ÏÇ¨Ìï≠ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.
    - ÏûÖÍ∏à ÌöüÏàò Ï∂îÏ†Å : depositÏù¥ Ïã§ÌñâÎê†ÎïåÎßàÎã§ ÏûÖÍ∏à ÌöüÏàòÎ•º Ï∂îÏ†ÅÌïòÎäî Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàòÍ∞Ä Ï∂îÍ∞ÄÎêòÏñ¥Ïïº Ìï©ÎãàÎã§.
    - Ïù¥Ïûê ÏßÄÍ∏âÏùÑ ÏúÑÌïú Î°úÏßÅ Ï∂îÍ∞Ä : ÏûÖÍ∏àÌöüÏàò 5ÌöåÎßàÎã§ ÏûîÏï°ÏùÑ Ï°∞ÌöåÌïòÍ≥†, Í∑∏ ÏûîÏï°(5ÌöåÏ∞® ÏûÖÍ∏àÏï° Ìè¨Ìï®)Ïùò 1%Î•º Ï∂îÍ∞ÄÎ°ú ÏûÖÍ∏àÌï¥Ïïº Ìï©ÎãàÎã§.
    - ÏßÄÍ∏âÌïú Ïù¥ÏûêÎ•º ÏûîÍ≥†Ïóê ÏóÖÎç∞Ïù¥Ìä∏ Ìï¥Ïïº Ìï©ÎãàÎã§.
- Í∞Å Í∏∞Îä•ÏóêÏÑú ÌïÑÏöîÌïú print Î¨∏ÎèÑ Ï†ÅÏ†àÌïòÍ≤å Ìè¨Ìï®ÌïòÏó¨ Ï∂îÍ∞ÄÌïòÍ≤†ÏäµÎãàÎã§.
    - Ïù¥Ïûê ÏßÄÍ∏â Ï°∞Í±¥ Ï∂©Ï°± Ïó¨Î∂Ä
    - Ïù¥Ïûê Í∏àÏï°(ÏÜåÏàòÏ†ê2ÏûêÎ¶¨)
- Î∂ÄÍ∞ÄÏ†ÅÏúºÎ°ú Ïù¥ÏûêÎ°ú Ïù∏Ìï¥ ÏÜåÏàòÏ†êÏù¥ ÎäòÏñ¥ÎÇòÎØÄÎ°ú, Ïù¥ÌõÑÏùò Í∏àÏï° Ï∂úÎ†•Í¥ÄÎ†® Î∂ÄÎ∂ÑÏùÄ ÏÜåÏàòÏ†ê 2ÏûêÎ¶¨ÍπåÏßÄÎßå Ï∂úÎ†•ÌïòÎäîÍ≤ÉÏúºÎ°ú ÏùºÍ¥Ñ ÏàòÏ†ïÌñàÏäµÎãàÎã§.

```python
import random # Í≥ÑÏ¢åÎ≤àÌò∏ ÎûúÎç§ ÏÉùÏÑ±Ïö©

class Account:
    # Í≥ÑÏ¢å Í≥ÑÏàò Ï∂úÎ†•ÏúÑÌïú Î≥ÄÏàò
    account_count = 0 # Î≥ÄÏàò Ï¥àÍ∏∞Ìôî

    # ÏÉùÏÑ±Ïûê ÏÉùÏÑ±
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # Í≥ÑÏ¢å ÏµúÏ¥à Í∞úÏÑ§Ïãú 0ÏõêÏù¥ÎØÄÎ°ú
        self.bank_name = 'SCÏùÄÌñâ' # ÏùÄÌñâ Ïù¥Î¶Ñ Í≥†Ï†ï
        self.account_number = self._generate_account_number()

        # ÏÉà Í≥ÑÏ¢å ÏÉùÏÑ±Ïãú ÎßàÎã§ Í≥ÑÏ¢å Í∞úÏàò +1
        Account.account_count += 1

        # Q7ÏóêÏÑú Ï∂îÍ∞ÄÎêú ÏΩîÎìú
        self.deposit_count = 0 # ÏûÖÍ∏à ÌöüÏàò Ï∂îÏ†Å ÏúÑÌïú Î≥ÄÏàò


    # ÎûúÎç§ Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Î©îÏÑúÎìú
    def _generate_account_number(self):
        """
        Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Í∑úÏπô : 3ÏûêÎ¶¨-2ÏûêÎ¶¨-6ÏûêÎ¶¨
        randomÌï®ÏàòÏùò randint Î©îÏÑúÎìú ÌôúÏö©
        Ïà´ÏûêÏùò Î≤îÏúÑÎäî ÏûêÎ¶¨ÏàòÏóê Îî∞ÎùºÏÑ§Ï†ï
        Íµ¨Ï°∞Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥¥Í∏∞ ÏúÑÌï¥ fstringÏÇ¨Ïö© (ÏùºÎ∞ò printÎ¨∏Ïù¥Î©¥ ""ÏôÄ ,Î•º ÏÑûÏñ¥Ïïº Ìï¥ÏÑú Í∞ÄÎèÖÏÑ± Îñ®Ïñ¥Ïßê)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # Í≥ÑÏ¢å Í∞úÏàò Ï∂úÎ†• Î©îÏÑúÎìú Ï∂îÍ∞Ä
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. ÏûÖÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    # Q7. Ïù¥ÏûêÏßÄÍ∏â Î©îÏÑúÎìúÎ•º depositÏùò ÌïòÏúÑ Î©îÏÑúÎìúÎ°ú Ï∂îÍ∞Ä
    def deposit(self, amount):
        # ÏûÖÍ∏àÏ†Ñ ÌòÑÏû¨ ÏûîÏï° Ï†ïÎ≥¥ Ï∂úÎ†•
        print("ÏûÖÍ∏à Ï†Ñ ÏûîÏï°:", "{:,}Ïõê".format(self.balance)) # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)

        # ÏµúÏÜå ÏûÖÍ∏àÏï° Í∏∞Ï§Ä Ï∂©Ï°± Ïó¨Î∂Ä ÌåêÎã®
        if amount >= 1:
            self.balance += amount # ÏûîÍ≥†Ïóê ÏûÖÍ∏àÏï° Ï∂îÍ∞Ä
            self.deposit_count += 1 # ÏûÖÍ∏àÌöüÏàò Ï¶ùÍ∞Ä
            # Í≤∞Í≥º Ï∂úÎ†•
            print(f"ÏûÖÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,.2f}Ïõê ÏûÖÎãàÎã§.") # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)


            # Q7. ÏûÖÍ∏à 5ÌöåÎãπ Ïù¥Ïûê ÏßÄÍ∏â Í∏∞Îä• Ï∂îÍ∞ÄÌïòÎäî ÏΩîÎìú
            if self.deposit_count % 5 == 0: # 5ÌöåÎãπ Ïù¥ÎØÄÎ°ú 5Î°ú ÎÇòÎàÑÏñ¥ ÎÇòÎ®∏ÏßÄÍ∞Ä 0Ïù¥ÎêòÏñ¥Ïïº Ìï®
                interest = self.balance * 0.01 # ÏûîÍ≥†(balance)Î•º Î∂àÎü¨Ïò§Í≥† Ìï¥Îãπ Í∏àÏï°Ïùò 1% Ïù¥ÏûêÎ°ú Í≥ÑÏÇ∞
                self.balance += interest # Ïù¥Ïûê ÏßÄÍ∏â Î∞è Î∞òÏòÅ
                print(f"ÎàÑÏ†Å ÏûÖÍ∏à ÏãúÎèÑÍ∞Ä {self.deposit_count}ÌöåÏûÖÎãàÎã§! ÎÇ¥Î∂ÄÏ†ïÏ±ÖÏóê Îî∞Îùº ÌòÑÏû¨ ÏûîÍ≥†Ïùò 1%Î•º Ïù¥ÏûêÎ°ú ÏßÄÍ∏âÌï©ÎãàÎã§.")
                print(f"Ï∂îÍ∞Ä Ïù¥Ïûê ÏßÄÍ∏â: {interest:,.2f}Ïõê - ÏÉàÎ°úÏö¥ ÏûîÏï°: {self.balance:,.2f}Ïõê")

        # ÏµúÏÜå ÏûÖÍ∏àÏï° ÎØ∏Ï∂©Ï°±Ïãú Î¨∏Íµ¨ Ï∂úÎ†•
        else:
            print(f"ÏûÖÍ∏à Ïã§Ìå®! - ÏûÖÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏµúÏÜå ÏûÖÍ∏àÏï°ÏùÄ 1ÏõêÏûÖÎãàÎã§. Ïû¨ÏûÖÍ∏àÏùÑ ÏãúÎèÑÌï¥Ï£ºÏÑ∏Ïöî!")


    # Q5. Ï∂úÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def withdraw(self, amount):
        # ÌòÑÏû¨ ÏûîÍ≥† Ï∂úÎ†•
        print("ÌòÑÏû¨ ÏûîÍ≥† :", "{:,}Ïõê".format(self.balance))

        # Ï∂úÍ≥† ÏöîÏ≤≠ Í∏àÏï°Ïóê ÎåÄÌïú Ï∂úÍ∏à Ïó¨Î∂Ä Í≤ÄÏ¶ù
        if amount <= self.balance:
            self.balance -= amount # Ï∂úÍ∏àÏöîÏ≤≠Í∏àÏï°ÏùÑ ÏûîÍ≥†Ïóê Î∞òÏòÅ(Ï∞®Í∞ê)
            print(f"Ï∂úÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,.2f}Ïõê ÏûÖÎãàÎã§.")
        else:
            print(f"Ï∂úÍ∏à Ïã§Ìå®! - Ï∂úÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏûîÏï°Ïù¥ Î∂ÄÏ°±Ìï©ÎãàÎã§.")


    # Q6. Ï†ïÎ≥¥Ï∂úÎ†• Î©îÏÑúÎìú Íµ¨ÌòÑÌïòÍ∏∞
    def display_info(self):
        # Í∞ÅÍ∞ÅÏùò Ï†ïÎ≥¥Î•º Ï∂úÎ†•ÌïòÎäî printÎ¨∏ ÏûëÏÑ±
        print(f"ÏùÄÌñâ Ïù¥Î¶Ñ: {self.bank_name}")
        print(f"ÏòàÍ∏àÏ£º: {self.account_name}")
        print(f"Í≥ÑÏ¢åÎ≤àÌò∏: {self.account_number}")
        print(f"ÏûîÏï°: {self.balance:,.2f}Ïõê")



# ÌÖåÏä§Ìä∏
customer1 = Account("ÍπÄÏäπÏàú") # Ïã†Í∑úÍ≥†Í∞ù ÏÉùÏÑ±

# 5Ìöå ÏûÖÍ∏à ÏãúÎèÑ
for _ in range(5):
    customer1.deposit(10000) # Îß§Î≤à ÎßåÏõêÏî© ÏûÖÍ∏à

# Í≥ÑÏ¢å Ï†ïÎ≥¥ Ï∂úÎ†•
customer1.display_info()
```

## Q10. ÏûÖÏ∂úÍ∏à ÎÇ¥Ïó≠ ÏΩîÎìú ÏóÖÎç∞Ïù¥Ìä∏ : ÏûÖÍ∏àÎÇ¥Ïó≠ÏùÑ Ï∂úÎ†•ÌïòÎäî deposit_history Î©îÏÑúÎìú, Ï∂úÍ∏àÎÇ¥Ïó≠ÏùÑ Ï∂úÎ†•ÌïòÎäî withdraw_history Î©îÏÑúÎìúÎ•º Ï∂îÍ∞Ä ÌïòÏÑ∏Ïöî.

- ÏûÖÏ∂úÍ∏à ÎÇ¥Ïó≠ÏùÑ Í∏∞Î°ùÌïòÏòÄÎã§Í∞Ä Ï∂úÎ†•ÌïòÎäî Î©îÏÑúÎìúÎ•º Í∞ÅÍ∞Å ÎßåÎì§Ïñ¥Ïïº Ìï©ÎãàÎã§.
- Ïù¥Î•º ÏúÑÌï¥ ÏûÖ/Ï∂úÍ∏àÏãú Î∞úÏÉùÌïòÎäî Í∞Å Í±∞Îûò ÎÇ¥Ïó≠ÏùÑ Ï†ÄÏû•Ìï† Î≥ÄÏàòÎ•º ÏÉùÏÑ±ÏûêÏóê Ï∂îÍ∞ÄÌï¥Ïïº Ìï©ÎãàÎã§.
- ÏûÖ/Ï∂úÍ∏àÏãúÎßàÎã§ Î∞úÏÉùÌïòÎäî Ï†ïÎ≥¥Î•º
- ÎòêÌïú Ìï¥Îãπ Ï†ïÎ≥¥Î•º Î∂àÎü¨ÏôÄÏÑú

```python
import random # Í≥ÑÏ¢åÎ≤àÌò∏ ÎûúÎç§ ÏÉùÏÑ±Ïö©

class Account:
    # Í≥ÑÏ¢å Í≥ÑÏàò Ï∂úÎ†•ÏúÑÌïú Î≥ÄÏàò
    account_count = 0 # Î≥ÄÏàò Ï¥àÍ∏∞Ìôî

    # ÏÉùÏÑ±Ïûê ÏÉùÏÑ±
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # Í≥ÑÏ¢å ÏµúÏ¥à Í∞úÏÑ§Ïãú 0ÏõêÏù¥ÎØÄÎ°ú
        self.bank_name = 'SCÏùÄÌñâ' # ÏùÄÌñâ Ïù¥Î¶Ñ Í≥†Ï†ï
        self.account_number = self._generate_account_number()

        # ÏÉà Í≥ÑÏ¢å ÏÉùÏÑ±Ïãú ÎßàÎã§ Í≥ÑÏ¢å Í∞úÏàò +1
        Account.account_count += 1

        # Q7ÏóêÏÑú Ï∂îÍ∞ÄÎêú ÏΩîÎìú
        self.deposit_count = 0 # ÏûÖÍ∏à ÌöüÏàò Ï∂îÏ†Å ÏúÑÌïú Î≥ÄÏàò

        # Q10. ÏóêÏÑú Ï∂îÍ∞ÄÎêú ÏΩîÎìú
        self.deposit_history_list = []  # ÏûÖÍ∏à ÎÇ¥Ïó≠ Ï†ÄÏû•ÏùÑ ÏúÑÌïú Î¶¨Ïä§Ìä∏
        self.withdraw_history_list = []  # Ï∂úÍ∏à ÎÇ¥Ïó≠ Ï†ÄÏû•ÏùÑ ÏúÑÌïú Î¶¨Ïä§Ìä∏


    # ÎûúÎç§ Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Î©îÏÑúÎìú
    def _generate_account_number(self):
        """
        Í≥ÑÏ¢åÎ≤àÌò∏ ÏÉùÏÑ± Í∑úÏπô : 3ÏûêÎ¶¨-2ÏûêÎ¶¨-6ÏûêÎ¶¨
        randomÌï®ÏàòÏùò randint Î©îÏÑúÎìú ÌôúÏö©
        Ïà´ÏûêÏùò Î≤îÏúÑÎäî ÏûêÎ¶¨ÏàòÏóê Îî∞ÎùºÏÑ§Ï†ï
        Íµ¨Ï°∞Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î≥¥Í∏∞ ÏúÑÌï¥ fstringÏÇ¨Ïö© (ÏùºÎ∞ò printÎ¨∏Ïù¥Î©¥ ""ÏôÄ ,Î•º ÏÑûÏñ¥Ïïº Ìï¥ÏÑú Í∞ÄÎèÖÏÑ± Îñ®Ïñ¥Ïßê)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # Í≥ÑÏ¢å Í∞úÏàò Ï∂úÎ†• Î©îÏÑúÎìú Ï∂îÍ∞Ä
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. ÏûÖÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    # Q7. Ïù¥ÏûêÏßÄÍ∏â Î©îÏÑúÎìúÎ•º depositÏùò ÌïòÏúÑ Î©îÏÑúÎìúÎ°ú Ï∂îÍ∞Ä
    def deposit(self, amount):
        # ÏûÖÍ∏àÏ†Ñ ÌòÑÏû¨ ÏûîÏï° Ï†ïÎ≥¥ Ï∂úÎ†•
        print("ÏûÖÍ∏à Ï†Ñ ÏûîÏï°:", "{:,}Ïõê".format(self.balance)) # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)

        # ÏµúÏÜå ÏûÖÍ∏àÏï° Í∏∞Ï§Ä Ï∂©Ï°± Ïó¨Î∂Ä ÌåêÎã®
        if amount >= 1:
            self.balance += amount # ÏûîÍ≥†Ïóê ÏûÖÍ∏àÏï° Ï∂îÍ∞Ä
            self.deposit_count += 1 # ÏûÖÍ∏àÌöüÏàò Ï¶ùÍ∞Ä
            self.deposit_history_list.append(amount)  # ÏûÖÍ∏à ÎÇ¥Ïó≠Ïóê Ï∂îÍ∞Ä
            # Í≤∞Í≥º Ï∂úÎ†•
            print(f"ÏûÖÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,.2f}Ïõê ÏûÖÎãàÎã§.") # Ï≤úÎã®ÏúÑ ÏâºÌëú Í∏∞Îä• Íµ¨ÌòÑ(Q6 ÎåÄÎπÑ)


            # Q7. ÏûÖÍ∏à 5ÌöåÎãπ Ïù¥Ïûê ÏßÄÍ∏â Í∏∞Îä• Ï∂îÍ∞ÄÌïòÎäî ÏΩîÎìú
            if self.deposit_count % 5 == 0: # 5ÌöåÎãπ Ïù¥ÎØÄÎ°ú 5Î°ú ÎÇòÎàÑÏñ¥ ÎÇòÎ®∏ÏßÄÍ∞Ä 0Ïù¥ÎêòÏñ¥Ïïº Ìï®
                interest = self.balance * 0.01 # ÏûîÍ≥†(balance)Î•º Î∂àÎü¨Ïò§Í≥† Ìï¥Îãπ Í∏àÏï°Ïùò 1% Ïù¥ÏûêÎ°ú Í≥ÑÏÇ∞
                self.balance += interest # Ïù¥Ïûê ÏßÄÍ∏â Î∞è Î∞òÏòÅ
                print(f"ÎàÑÏ†Å ÏûÖÍ∏à ÏãúÎèÑÍ∞Ä {self.deposit_count}ÌöåÏûÖÎãàÎã§! ÎÇ¥Î∂ÄÏ†ïÏ±ÖÏóê Îî∞Îùº ÌòÑÏû¨ ÏûîÍ≥†Ïùò 1%Î•º Ïù¥ÏûêÎ°ú ÏßÄÍ∏âÌï©ÎãàÎã§.")
                print(f"Ï∂îÍ∞Ä Ïù¥Ïûê ÏßÄÍ∏â: {interest:,.2f}Ïõê - ÏÉàÎ°úÏö¥ ÏûîÏï°: {self.balance:,.2f}Ïõê")

        # ÏµúÏÜå ÏûÖÍ∏àÏï° ÎØ∏Ï∂©Ï°±Ïãú Î¨∏Íµ¨ Ï∂úÎ†•
        else:
            print(f"ÏûÖÍ∏à Ïã§Ìå®! - ÏûÖÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏµúÏÜå ÏûÖÍ∏àÏï°ÏùÄ 1ÏõêÏûÖÎãàÎã§. Ïû¨ÏûÖÍ∏àÏùÑ ÏãúÎèÑÌï¥Ï£ºÏÑ∏Ïöî!")


    # Q5. Ï∂úÍ∏à Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def withdraw(self, amount):
        # ÌòÑÏû¨ ÏûîÍ≥† Ï∂úÎ†•
        print("ÌòÑÏû¨ ÏûîÍ≥† :", "{:,}Ïõê".format(self.balance))

        # Ï∂úÍ≥† ÏöîÏ≤≠ Í∏àÏï°Ïóê ÎåÄÌïú Ï∂úÍ∏à Ïó¨Î∂Ä Í≤ÄÏ¶ù
        if amount <= self.balance:
            self.balance -= amount # Ï∂úÍ∏àÏöîÏ≤≠Í∏àÏï°ÏùÑ ÏûîÍ≥†Ïóê Î∞òÏòÅ(Ï∞®Í∞ê)
            # Q10. Ï∂úÍ∏à ÎÇ¥Ïó≠ historyÎ°ú Ï†ÄÏû•
            self.withdraw_history_list.append(amount)
            print(f"Ï∂úÍ∏à ÏÑ±Í≥µ! - ÌòÑÏû¨ ÏûîÏï°ÏùÄ {self.balance:,.2f}Ïõê ÏûÖÎãàÎã§.")
        else:
            print(f"Ï∂úÍ∏à Ïã§Ìå®! - Ï∂úÍ∏à ÏãúÎèÑÏï°: {amount}Ïõê. ÏûîÏï°Ïù¥ Î∂ÄÏ°±Ìï©ÎãàÎã§.")


    # Q6. Ï†ïÎ≥¥Ï∂úÎ†• Î©îÏÑúÎìú Íµ¨ÌòÑÌïòÍ∏∞
    def display_info(self):
        # Í∞ÅÍ∞ÅÏùò Ï†ïÎ≥¥Î•º Ï∂úÎ†•ÌïòÎäî printÎ¨∏ ÏûëÏÑ±
        print(f"ÏùÄÌñâ Ïù¥Î¶Ñ: {self.bank_name}")
        print(f"ÏòàÍ∏àÏ£º: {self.account_name}")
        print(f"Í≥ÑÏ¢åÎ≤àÌò∏: {self.account_number}")
        print(f"ÏûîÏï°: {self.balance:,.2f}Ïõê")


    # Q10. ÏûÖÍ∏à ÎÇ¥Ïó≠ Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def deposit_history(self):
        for deposit in self.deposit_history_list: # ÏÉùÏÑ±ÏûêÏóê ÏûàÎäî deposit_history_listÎ•º iterationÌïòÎ©∞ Ï†ïÎ≥¥ Ï°∞Ìöå
            print(f" - {deposit:,.2f}Ïõê ÏûÖÍ∏à") # Í∞Å ÏûÖÍ∏à Í∏∞Î°ùÏùÑ Ï∂úÎ†•
        if not self.deposit_history_list: # ÏïÑÏßÅ Í∏∞Î°ùÏù¥ ÏóÜÎäî Í≤ΩÏö∞
            print("ÏûÖÍ∏à ÎÇ¥Ïó≠Ïù¥ ÏóÜÏäµÎãàÎã§.")

    # Q10. Ï∂úÍ∏à ÎÇ¥Ïó≠ Î©îÏÑúÎìú Ï∂îÍ∞Ä
    def withdraw_history(self):
        for withdraw in self.withdraw_history_list: # ÏÉùÏÑ±ÏûêÏóê ÏûàÎäî withdraw_history_listÎ•º iterationÌïòÎ©∞ Ï†ïÎ≥¥ Ï°∞Ìöå
            print(f" - {withdraw:,.2f}Ïõê Ï∂úÎ†•") # Í∞Å Ï∂úÍ∏à Í∏∞Î°ùÏùÑ Ï∂úÎ†•
        if not self.withdraw_history_list: # ÏïÑÏßÅ Í∏∞Î°ùÏù¥ ÏóÜÎäî Í≤ΩÏö∞
            print("Ï∂úÍ∏à ÎÇ¥Ïó≠Ïù¥ ÏóÜÏäµÎãàÎã§.")


# ÌÖåÏä§Ìä∏
customer1 = Account("ÍπÄÏäπÏàú")
customer1.deposit(10000)  # ÏûÖÍ∏à
customer1.withdraw(5000)  # Ï∂úÍ∏à
customer1.deposit(20000)  # ÏûÖÍ∏à

# ÏûÖÍ∏à Î∞è Ï∂úÍ∏à ÎÇ¥Ïó≠ Ï∂úÎ†•
customer1.deposit_history()
customer1.withdraw_history()
```

## **Q6. Ï§ëÎèÑÌè¨Í∏∞Ïûê Í¥ÄÎ¶¨Î•º ÏúÑÌï¥ Í∑∏Î£®Ï†ïÎ≥¥Î•º ÏÇ≠Ï†úÌïòÎäî Î©îÏÑúÎìúÎ•º Ï∂îÍ∞ÄÌï¥Î≥¥ÏÑ∏Ïöî. Ìï¥Îãπ Î©îÏÑúÎìúÎ•º Ìò∏Ï∂úÌïòÎ©¥ Î™ÖÎã®Ïù¥ Í∞±Ïã†ÎêòÏñ¥Ïïº Ìï©ÎãàÎã§. ÌòÑÏû¨ Î™ÖÎã®ÏóêÏÑú ÍπÄÍ≤ΩÎØº, ÍπÄÏòÅÏã† ÎãòÏùÑ ÏÇ≠Ï†úÌïú ÌõÑ Í∞±Ïã†Îêú Î™ÖÎã®ÏùÑ Ï∂úÎ†•ÌïòÏÑ∏Ïöî.**

- Í∏∞Îä• Íµ¨ÌòÑÏùÑ ÏúÑÌï¥ remove_member ÎùºÎäî ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÎ•º ÏÉùÏÑ±Ìï¥Ïïº Ìï©ÎãàÎã§.
- instances ÎÇ¥Ïóê Ï†ÄÏû•Îêú ÌïôÏÉùÏ†ïÎ≥¥Î•º Í∞ÄÏßÄÍ≥† Ï†ëÍ∑ºÌï¥ÏÑú ÌäπÏ†ï Ï†ïÎ≥¥ÎßåÏùÑ ÏÇ≠Ï†úÌïòÎèÑÎ°ù Ìï©ÎãàÎã§.
- ÏÇ≠Ï†úÌï† Ï†ïÎ≥¥Í∞Ä ÏóÜÎäî Í≤ΩÏö∞ Ïò§Î•òÎ©îÏãúÏßÄÍ∞Ä Îú®ÎèÑÎ°ù ÌñàÏäµÎãàÎã§.

```python
import pandas as pd # csv ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†•Ïö©

class Aiffel:
    total_id_count = 0  # ÏÉùÏÑ±Îêú ÌïôÎ≤àÏùò Ï¥ù Í∞úÏàò
    instances = {}  # ÏÉùÏÑ±Îêú Aiffel Ïù∏Ïä§ÌÑ¥Ïä§ Î™©Î°ùÏùÑ ÌÅ¥ÎûòÏä§ Î†àÎ≤®ÏóêÏÑú Í¥ÄÎ¶¨
    members_instances = {} # ÏΩîÏñ¥ Î™ÖÎã® dictÎ•º Î∞îÌÉïÏúºÎ°ú Aiffel ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Î•º ÏÉùÏÑ±Ïö© dict

    def __init__(self, name, admission_year="2023", term=29, course_type="ÏΩîÏñ¥"):
        self.name = name  # ÌïôÏÉùÏùò Ïù¥Î¶Ñ
        self.admission_year = admission_year  # ÏûÖÌïô Ïó∞ÎèÑ
        self.term = term  # Í∏∞Ïàò
        self.course_type = course_type  # Í≥ºÏ†ï Ï¢ÖÎ•ò
        self.student_id = self.generate_student_id()  # ÌïôÎ≤à ÏÉùÏÑ± Î©îÏÑúÎìú Ìò∏Ï∂ú
        self.score_points = 0  # Í∞úÎ≥Ñ Í∑∏Î£®Ïùò ÌÄòÏä§Ìä∏ Ï†êÏàò
        self.penalty_points = 0  # Í∞úÎ≥Ñ Í∑∏Î£®Ïùò ÌÄòÏä§Ìä∏ penalty
        Aiffel.instances[name] = self  # ÏÉùÏÑ±Îêú Ïù∏Ïä§ÌÑ¥Ïä§Î•º Ïù¥Î¶ÑÏùÑ keyÎ°ú ÎîïÏÖîÎÑàÎ¶¨Ïóê Ï∂îÍ∞Ä
        Aiffel.total_id_count += 1  # ÌïôÎ≤à ÏÉùÏÑ±ÏãúÎßàÎã§ Ï¥ù ÌïôÎ≤à Í∞úÏàò Ï¶ùÍ∞Ä

    def generate_student_id(self):
        year = self.admission_year[-2:]  # ÏûÖÌïô Ïó∞ÎèÑÏùò ÎßàÏßÄÎßâ Îëê ÏûêÎ¶¨
        course_type_num = '2' if self.course_type == "ÏΩîÏñ¥" else '1'  # Í≥ºÏ†ï Ï¢ÖÎ•òÏóê Îî∞Î•∏ Ïà´Ïûê Ìï†Îãπ
        student_number = str(Aiffel.total_id_count + 1).zfill(2)  # ÌïôÎ≤àÏùò ÏàúÏÑúÎ•º ÎÇòÌÉÄÎÇ¥Îäî Îëê ÏûêÎ¶¨ Ïà´Ïûê
        student_id = f"{year}{str(self.term).zfill(2)}{course_type_num}{student_number}"  # ÌïôÎ≤à Ï°∞Ìï©
        return student_id  # ÏÉùÏÑ±Îêú ÌïôÎ≤à Î∞òÌôò

    @classmethod
    def get_id_num(cls):
        return cls.total_id_count

    # Q3. ÏóêÏÑú Ï∂îÍ∞Ä
    def score(self, points):
        if points in [1, 2, 3]:
            self.score_points += points  # Ï°∞Í±¥Ïóê Ìï¥ÎãπÌï† Í≤ΩÏö∞ Ï†êÏàò Ï∂îÍ∞Ä
        else:
            print("Ï†êÏàòÎäî 1Ï†ê, 2Ï†ê, 3Ï†ê Ï§ë 1Í∞úÎßå ÏûÖÎ†• Í∞ÄÎä•Ìï©ÎãàÎã§.")

    # Q4. ÏóêÏÑú Ï∂îÍ∞Ä
    def penalty(self, points):
        self.penalty_points -= points  # ÏûÖÎ†•Î∞õÏùÄ Ï†êÏàòÎ•º Ï∞®Í∞ê

    # Q5. ÏóêÏÑú Ï∂îÍ∞Ä
    @classmethod
    def display_info(cls):
        data = {
            "Ïù¥Î¶Ñ": [],
            "ÌïôÎ≤à": [],
            "Í≥ºÏ†ï Ï¢ÖÎ•ò": [],
            "ÌÄòÏä§Ìä∏ Ï†êÏàò Ï¥ùÏ†ê": [],
            "ÌÄòÏä§Ìä∏ ÏûÖÎ†• Ï¥ùÏ†ê": [],
            "ÌÄòÏä§Ìä∏ penalty Ï¥ùÌï©": []
        }

        # ÌÅ¥ÎûòÏä§ ÎÇ¥Î∂ÄÏóêÏÑú instances Î≥ÄÏàòÏóê Ï†ëÍ∑º
        for name in cls.instances.keys():
            student = cls.instances[name]
            data["Ïù¥Î¶Ñ"].append(student.name)
            data["ÌïôÎ≤à"].append(student.student_id)
            data["Í≥ºÏ†ï Ï¢ÖÎ•ò"].append(student.course_type)
            data["ÌÄòÏä§Ìä∏ Ï†êÏàò Ï¥ùÏ†ê"].append(student.score_points + student.penalty_points)
            data["ÌÄòÏä§Ìä∏ ÏûÖÎ†• Ï¥ùÏ†ê"].append(student.score_points)
            data["ÌÄòÏä§Ìä∏ penalty Ï¥ùÌï©"].append(student.penalty_points)

        df = pd.DataFrame(data)
        display(df)

    # Q6. Ï§ëÎèÑÌè¨Í∏∞Ïûê Í¥ÄÎ¶¨Î•º ÏúÑÌïú Î©îÏÑúÎìú Ï∂îÍ∞Ä - Î™ÖÎã® ÏÇ≠Ï†ú
    @classmethod
    def remove_member(cls, name):
        if name in cls.instances:
            del cls.instances[name]  # Ìï¥Îãπ Í∑∏Î£® Ï†ïÎ≥¥ ÏÇ≠Ï†ú
            print(f"{name} ÎãòÏùò Ï†ïÎ≥¥Í∞Ä ÏÇ≠Ï†úÎêòÏóàÏäµÎãàÎã§.")
        else:
            print(f"{name} ÎãòÏùÄ Î™ÖÎã®Ïóê Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏäµÎãàÎã§.")


# ÌÅ¥ÎûòÏä§ÎÇ¥Ïóê Í∑∏Î£® Ï†ïÎ≥¥ ÏÉùÏÑ± Î∞è Ï†ÄÏû•
# ÏΩîÏñ¥ Î™ÖÎã® dict
core_members =  {'Í∞ïÏòÅÌòÑ': 0, 'ÍπÄÍ≤ΩÎØº': 0, 'ÍπÄÎØºÍ∏∞': 0, 'ÍπÄÏäπÏàú': 0, 'ÍπÄÏñëÌù¨': 0, 'ÍπÄÏòÅÏã†': 0,
                 'Ïò§Ïö∞ÏßÑ': 0, 'Ïú§ÌòúÏú§': 0, 'Ïù¥ÏÉÅÏõê': 0, 'Ïù¥ÏÑ†Ïû¨': 0, 'Ïù¥Ïä¨': 0, 'Ïù¥ÏäπÌôò': 0, 'Ï†ÑÎØºÍ∑ú': 0,
                 'Ï°∞Î≥¥Í≤∏': 0, 'Ï°∞ÏàòÎØº': 0, 'ÏµúÍ∞ïÌõà': 0, 'ÌïúÌòÑÏ¢Ö': 0, 'ÌôçÏ±ÑÎ¶º': 0}

# ÏΩîÏñ¥ Î™ÖÎã®ÏùÑ Î∞îÌÉïÏúºÎ°ú Aiffel Ïù∏Ïä§ÌÑ¥Ïä§ ÏÉùÏÑ±

# Í∞Å ÌïôÏÉùÏùò Ïù¥Î¶ÑÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Aiffel ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Î•º ÏÉùÏÑ±
# Aiffel ÌÅ¥ÎûòÏä§Ïùò ÏÉùÏÑ±ÏûêÏóê Ï†ïÏùòÎêú Í∏∞Î≥∏Í∞í(Ïó∞ÎèÑ, Í∏∞Ïàò, Í≥ºÏ†ï Ï¢ÖÎ•ò) Ï†ÅÏö©
for name in core_members.keys():
    members_instances[name] = Aiffel(name)

# ÌïôÏÉù Ï†ïÎ≥¥ Ï°∞Ìöå
Aiffel.display_info()

# remove_memberÌÖåÏä§Ìä∏
# ÍπÄÍ≤ΩÎØºÍ≥º ÍπÄÏòÅÏã†ÏùÑ ÏÇ≠Ï†ú
Aiffel.remove_member("ÍπÄÍ≤ΩÎØº")
Aiffel.remove_member("ÍπÄÏòÅÏã†")

# Í∞±Ïã†Îêú Î™ÖÎã® Ï∂úÎ†•
Aiffel.display_info()

# Îã§ÏãúÌïúÎ≤à ÏÇ≠Ï†ú ÏãúÎèÑ
Aiffel.remove_member("ÍπÄÏòÅÏã†")
```

## **Q7. Aiffel ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜçÌïòÏó¨, Í∑∏Î£® Î™ÖÎã®ÏùÑ Î∞îÌÉïÏúºÎ°ú 2Ïù∏, 3Ïù∏, 4Ïù∏ÏúºÎ°ú ÎûúÎç§ Í∑∏Î£πÌïëÏùÑ Ìï¥Ï£ºÎäî Î©îÏÑúÎìúÍ∞Ä Ï∂îÍ∞ÄÎêú Aiffel_Group ÌÅ¥ÎûòÏä§Î•º ÎßåÎì§Ïñ¥ Î≥¥ÏÑ∏Ïöî. Í∑∏Î£πÏù¥Î¶ÑÏùÄ Ï¥ù 8Í∞úÏûÖÎãàÎã§. (ÏÉÅÏÉù, Ïã†Î¢∞, Ïó¥Ï†ï, Ïù¥ÎÅî, ÏÑ±Ïû•, Í∞ÄÏπò, Í≥µÏú†, ÌôïÏÇ∞)**

- Aiffel ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜçÌïòÎ©∞ Î™®Îì† Ï†ïÎ≥¥Î•º Îã§ ÏäπÍ≥Ñ Î∞õÏïÑÏïº Ìï©ÎãàÎã§.
- ÎûúÎç§ Í∑∏Î£®ÌïëÏùÑ ÏúÑÌïú Î©îÏÑúÎìúÍ∞Ä Ï∂îÍ∞ÄÎêòÏñ¥Ïïº Ìï©ÎãàÎã§.
    - create_random_groupeÎùºÎäî Î©îÏÑúÎìúÎ•º Ï∂îÍ∞ÄÌïòÍ≤†ÏäµÎãàÎã§.
    - Ìï¥Îãπ Î©îÏÑúÎìúÎäî Ï†ÑÏ≤¥ Í∑∏Î£® Ïù∏ÏõêÏ†ïÎ≥¥ÏóêÏÑú 2Ïù∏, 3Ïù∏, 4Ïù∏Ïùò Ï°∞Ìï©ÏùÑ ÎûúÎç§ÏúºÎ°ú ÌïòÏó¨ 8Í∞úÏùò Í∑∏Î£πÏóê ÎûúÎç§ÏúºÎ°ú Î∞∞ÏπòÌï©ÎãàÎã§.
    - Ï¶â, Í∑∏Î£πÏóê Î∞∞ÏπòÌï† Ïù∏ÏõêÏùÑ ÎΩëÎäî Í≤ÉÎèÑ ÎûúÎç§, Ïñ¥Îäê Í∑∏Î£πÏóê Î∞∞ÏπòÌï†ÏßÄÎèÑ ÎûúÎç§Ïù¥ ÎêòÏñ¥Ïïº Ìï©ÎãàÎã§.
    - Îã§Îßå, Î∞∞ÏπòÎê† Í∑∏Î£πÏùÄ 8Í∞úÍ∞Ä Î™®Îëê Ï≤òÏùåÎ∂ÄÌÑ∞ ÏÉùÏÑ±ÎêòÏñ¥ ÏûàÏñ¥Ïïº Ìï©ÎãàÎã§.
- ÏÉùÏÑ±Îêú Í∑∏Î£π Ï†ïÎ≥¥Îäî ÌÅ¥ÎûòÏä§Ïùò Î≥ÄÏàòÎ°ú Ï†ÄÏû•Ìï¥ÏÑú Í¥ÄÎ¶¨Ìï©ÎãàÎã§.

```python
import random

class Aiffel_Group(Aiffel):
    # Í∑∏Î£π Ïù¥Î¶ÑÏùÑ ÎØ∏Î¶¨ Ï†ïÏùò
    group_names = ["ÏÉÅÏÉù", "Ïã†Î¢∞", "Ïó¥Ï†ï", "Ïù¥ÎÅî", "ÏÑ±Ïû•", "Í∞ÄÏπò", "Í≥µÏú†", "ÌôïÏÇ∞"]
    # ÏÉùÏÑ±Îêú Í∑∏Î£πÏùÑ Ï†ÄÏû•Ìï† ÎîïÏÖîÎÑàÎ¶¨Î•º Ï¥àÍ∏∞Ìôî
    groups = {}

    @classmethod
    def create_random_groups(cls):
        # Aiffel ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Îì§(ÌïôÏÉùÎì§)Ïùò Ïù¥Î¶ÑÏùÑ Î¶¨Ïä§Ìä∏Î°ú Í∞ÄÏ†∏Ïò§Í∏∞
        all_students = list(cls.instances.keys())
        # ÌïôÏÉùÎì§Ïùò ÏàúÏÑúÎ•º ÎûúÎç§ÏúºÎ°ú ÏÑûÍ∏∞
        random.shuffle(all_students)


        # Î®ºÏ†Ä 8Í∞úÏùò Í∑∏Î£πÏùÑ ÏÉùÏÑ±
        for group_name in cls.group_names:
            cls.groups[group_name] = []



        # Î™®Îì† ÌïôÏÉùÏùÑ Í∑∏Î£πÏóê ÎûúÎç§ÏúºÎ°ú Î∞∞Ïπò
        while all_students:
            for group_name in cls.group_names:
                if not all_students:
                    break
                # Í∞Å Í∑∏Î£πÏóê 2Î™Ö, 3Î™Ö ÎòêÎäî 4Î™ÖÏùò ÌïôÏÉùÏùÑ ÎûúÎç§ÏúºÎ°ú Î∞∞Ïπò
                group_size = random.choice([2, 3, 4])
                # ÎÇ®ÏùÄ ÌïôÏÉù ÏàòÍ∞Ä Í≤∞Ï†ïÎêú Í∑∏Î£π ÌÅ¨Í∏∞Î≥¥Îã§ ÏûëÏúºÎ©¥, ÎÇ®ÏùÄ ÌïôÏÉù ÏàòÎ•º Í∑∏Î£π ÌÅ¨Í∏∞Î°ú ÏÑ§Ï†ï
                group_size = min(group_size, len(all_students))

                # ÌïôÏÉùÎì§ÏùÑ Í∑∏Î£πÏóê Ìï†Îãπ
                cls.groups[group_name].extend(all_students[:group_size])
                all_students = all_students[group_size:]

    @classmethod
    def display_groups(cls):
        # Ï†ÄÏû•Îêú Î™®Îì† Í∑∏Î£πÍ≥º Í∑∏ Íµ¨ÏÑ±ÏõêÎì§ÏùÑ Ï∂úÎ†•
        for group_name, members in cls.groups.items():
            print(f"Í∑∏Î£π Ïù¥Î¶Ñ: {group_name}, Íµ¨ÏÑ±Ïõê: {', '.join(members)}")

# Aiffel_Group ÌÅ¥ÎûòÏä§Ïùò ÎûúÎç§ Í∑∏Î£π ÏÉùÏÑ± Î∞è Ï∂úÎ†• Î©îÏÑúÎìúÎ•º Ìò∏Ï∂ú
Aiffel_Group.create_random_groups()
Aiffel_Group.display_groups()
```

    - Í∑∏Î£π Ïù¥Î¶ÑÍ≥º Ï†êÏàòÎ•º Ïù∏ÏûêÎ°ú Î∞õÏïÑ ÏûëÎèôÌï©ÎãàÎã§.
    - ÏßÄÏ†ïÎêú Í∑∏Î£πÏóê ÏÜçÌïú Î©§Î≤ÑÎì§Ïùò Í∏∏ÎìúÍ∞Ä Î¨¥ÏóáÏù∏ÏßÄ ÌôïÏù∏ÌïòÎäî Í∏∞Îä•Ïù¥ ÏûàÏñ¥Ïïº Ìï©ÎãàÎã§.

```python
class Aiffel_Group(Aiffel):
    # Í∑∏Î£π Ïù¥Î¶ÑÏùÑ ÎØ∏Î¶¨ Ï†ïÏùò
    group_names = ["ÏÉÅÏÉù", "Ïã†Î¢∞", "Ïó¥Ï†ï", "Ïù¥ÎÅî", "ÏÑ±Ïû•", "Í∞ÄÏπò", "Í≥µÏú†", "ÌôïÏÇ∞"]
    # ÏÉùÏÑ±Îêú Í∑∏Î£πÏùÑ Ï†ÄÏû•Ìï† ÎîïÏÖîÎÑàÎ¶¨Î•º Ï¥àÍ∏∞Ìôî
    groups = {}

    @classmethod
    def create_random_groups(cls):
        # Aiffel ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Îì§(ÌïôÏÉùÎì§)Ïùò Ïù¥Î¶ÑÏùÑ Î¶¨Ïä§Ìä∏Î°ú Í∞ÄÏ†∏Ïò§Í∏∞
        all_students = list(cls.instances.keys())
        # ÌïôÏÉùÎì§Ïùò ÏàúÏÑúÎ•º ÎûúÎç§ÏúºÎ°ú ÏÑûÍ∏∞
        random.shuffle(all_students)

        # Ï†ïÏùòÎêú Í∑∏Î£π Ïù¥Î¶ÑÎßåÌÅº Î∞òÎ≥µÌïòÏó¨ Í∞Å Í∑∏Î£πÏóê ÌïôÏÉùÎì§ÏùÑ Ìï†Îãπ
        for group_name in cls.group_names:
            # Í∑∏Î£πÏùò Íµ¨ÏÑ±Ïõê ÏàòÎ•º 2, 3, 4 Ï§ë ÎûúÎç§ÏúºÎ°ú Í≤∞Ï†ï
            group_size = random.choice([2, 3, 4])
            if group_size > len(all_students):
                group_size = len(all_students) # ÎßåÏïΩ ÎÇ®ÏùÄ ÌïôÏÉù ÏàòÍ∞Ä Í≤∞Ï†ïÎêú Í∑∏Î£π ÌÅ¨Í∏∞Î≥¥Îã§ ÏûëÏúºÎ©¥ ÎÇ®ÏùÄ ÌïôÏÉù ÏàòÎ•º Í∑∏Î£π ÌÅ¨Í∏∞Î°ú ÏÑ§Ï†ï

            # Í≤∞Ï†ïÎêú Í∑∏Î£π ÌÅ¨Í∏∞ÎßåÌÅº ÌïôÏÉùÏùÑ Í∑∏Î£πÏóê Ìï†Îãπ
            group_members = all_students[:group_size]
            # Ìï†ÎãπÎêú ÌïôÏÉùÎì§ÏùÑ Ï†ÑÏ≤¥ ÌïôÏÉù Î™©Î°ùÏóêÏÑú Ï†úÍ±∞
            all_students = all_students[group_size:]

            # Í∑∏Î£πÎ™ÖÍ≥º Íµ¨ÏÑ±ÏõêÏùÑ Í∑∏Î£π ÎîïÏÖîÎÑàÎ¶¨Ïóê Ï∂îÍ∞Ä
            cls.groups[group_name] = group_members

            # Î™®Îì† ÌïôÏÉùÏù¥ Í∑∏Î£πÏóê Ìï†ÎãπÎêòÎ©¥ Î∞òÎ≥µÏùÑ Ï§ëÎã®
            if not all_students:
                break

    @classmethod
    def display_groups(cls):
        # Ï†ÄÏû•Îêú Î™®Îì† Í∑∏Î£πÍ≥º Í∑∏ Íµ¨ÏÑ±ÏõêÎì§ÏùÑ Ï∂úÎ†•
        for group_name, members in cls.groups.items():
            print(f"Í∑∏Î£π Ïù¥Î¶Ñ: {group_name}, Íµ¨ÏÑ±Ïõê: {', '.join(members)}")

    # Q 10. ÏóêÏÑú Ï∂îÍ∞Ä
    @classmethod
    def group2guild_score(cls, group_name, score):
        """
        ÏßÄÏ†ïÎêú Í∑∏Î£πÏóê ÏÜçÌïú Î©§Î≤ÑÏùò Í∏∏ÎìúÎ•º ÌôïÏù∏ÌïòÍ≥† Ï†êÏàòÎ•º Î≥ÄÍ≤Ω
        ÏßÄÏ†ïÎêú Í∑∏Î£πÏóê Ï†êÏàòÍ∞Ä Ï∂îÍ∞ÄÎêòÎ©¥ Ìï¥Îãπ Í∑∏Î£πÏóê ÏÜçÌïú Î©§Î≤ÑÎì§Ïùò Í∏∏ÎìúÏóêÎèÑ ÎèôÏùºÌïú Ï†êÏàòÍ∞Ä Ï∂îÍ∞Ä
        """
        # ÏßÄÏ†ïÎêú Í∑∏Î£π Ïù¥Î¶ÑÏù¥ cls.groupsÏóê Ï°¥Ïû¨ÌïòÎäîÏßÄ ÌôïÏù∏
        if group_name in cls.groups:
            # Í∑∏Î£πÏóê ÏÜçÌïú Î™®Îì† Î©§Î≤Ñ(ÌïôÏÉù)Î•º ÏàúÌöåÌïòÎ©∞ ÌôïÏù∏
            for student in cls.groups[group_name]:
                # Í∞Å ÌïôÏÉùÏù¥ ÏÜçÌïú Í∏∏ÎìúÎ•º Ï∞æÍ∏∞ ÏúÑÌï¥ Î™®Îì† Í∏∏ÎìúÎ•º ÏàúÌöåÌïòÎ©∞ ÌôïÏù∏
                for guild_name, members in Aiffel_Guild.guilds.items():
                    # Ìï¥Îãπ ÌïôÏÉùÏù¥ Í∏∏ÎìúÏùò Î©§Î≤Ñ Î™©Î°ùÏóê ÏûàÎäîÏßÄÏóê Îî∞Îùº Ï†êÏàò Ï∂îÍ∞Ä
                    if student in members:
                        # ÌïôÏÉùÏù¥ ÏÜçÌïú Í∏∏ÎìúÏóê Ï†êÏàòÎ•º Ï∂îÍ∞Ä
                        Aiffel_Guild.guild_score(guild_name, score)
                        """
                        Í∑∏Î£® 1Î™Ö = ÏÜåÏÜç Í∏∏Îìú 1Í∞ú -> ÏÜåÏÜç Í∏∏ÎìúÎ•º Ï∞æÏïòÏúºÎØÄÎ°ú Ï†êÏàòÎ•º Ï†ÅÏö©ÌïòÍ≥† Ï¢ÖÎ£å
                        """
                        break

    @classmethod
    def display_guilds(cls): # Ï†ÄÏû•Îêú Î™®Îì† Í∏∏ÎìúÏôÄ Í∑∏ Íµ¨ÏÑ±ÏõêÎì§ÏùÑ Ï∂úÎ†•ÌïòÎäî Î©îÏÑúÎìú
        for guild_name, members in cls.guilds.items():
            print(f"Í∏∏Îìú Ïù¥Î¶Ñ: {guild_name}, Íµ¨ÏÑ±Ïõê: {', '.join(members)}")



# Í≤∞Í≥º ÌôïÏù∏

# Í∏∏Îìú ÏÉùÏÑ± Î∞è Ï¥àÍ∏∞ Í∏∏Îìú Ï†ïÎ≥¥ ÌëúÏãú
Aiffel_Guild.create_random_guilds()
Aiffel_Guild.display_guilds()

# ÎûúÎç§ Í∑∏Î£π ÏÉùÏÑ± Î∞è Ï¥àÍ∏∞ Í∑∏Î£π Ï†ïÎ≥¥ ÌëúÏãú
Aiffel_Group.create_random_groups()
Aiffel_Group.display_groups()

# Ï¥àÍ∏∞ Í∑∏Î£® Ï†êÏàò Ï†ïÎ≥¥ ÌëúÏãú
Aiffel.display_info()

# 'ÏÉÅÏÉÅ' Í∑∏Î£πÏóê +2 Ï†êÏàò Î∂ÄÏó¨
Aiffel_Group.group2guild_score("ÏÉÅÏÉù", 2)

# Ï†êÏàò Î∂ÄÏó¨ ÌõÑ Í∏∏Îìú Î∞è Í∑∏Î£® Ï†êÏàò Ï†ïÎ≥¥ Ïû¨ÌëúÏãú
Aiffel.display_info()
```

```python
"""
Aiffel_Group Ïùò Ï†ïÎ≥¥ Ïò§Î≤ÑÎùºÏù¥Îî©
"""
class Aiffel_Group(Aiffel):
    # Í∑∏Î£π Ïù¥Î¶ÑÏùÑ ÎØ∏Î¶¨ Ï†ïÏùò
    group_names = ["ÏÉÅÏÉù", "Ïã†Î¢∞", "Ïó¥Ï†ï", "Ïù¥ÎÅî", "ÏÑ±Ïû•", "Í∞ÄÏπò", "Í≥µÏú†", "ÌôïÏÇ∞"]
    # ÏÉùÏÑ±Îêú Í∑∏Î£πÏùÑ Ï†ÄÏû•Ìï† ÎîïÏÖîÎÑàÎ¶¨Î•º Ï¥àÍ∏∞Ìôî
    groups = {}

    @classmethod
    def create_random_groups(cls):
        # Aiffel ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Îì§(ÌïôÏÉùÎì§)Ïùò Ïù¥Î¶ÑÏùÑ Î¶¨Ïä§Ìä∏Î°ú Í∞ÄÏ†∏Ïò§Í∏∞
        all_students = list(cls.instances.keys())
        # ÌïôÏÉùÎì§Ïùò ÏàúÏÑúÎ•º ÎûúÎç§ÏúºÎ°ú ÏÑûÍ∏∞
        random.shuffle(all_students)

        # Ï†ïÏùòÎêú Í∑∏Î£π Ïù¥Î¶ÑÎßåÌÅº Î∞òÎ≥µÌïòÏó¨ Í∞Å Í∑∏Î£πÏóê ÌïôÏÉùÎì§ÏùÑ Ìï†Îãπ
        for group_name in cls.group_names:
            # Í∑∏Î£πÏùò Íµ¨ÏÑ±Ïõê ÏàòÎ•º 2, 3, 4 Ï§ë ÎûúÎç§ÏúºÎ°ú Í≤∞Ï†ï
            group_size = random.choice([2, 3, 4])
            if group_size > len(all_students):
                group_size = len(all_students) # ÎßåÏïΩ ÎÇ®ÏùÄ ÌïôÏÉù ÏàòÍ∞Ä Í≤∞Ï†ïÎêú Í∑∏Î£π ÌÅ¨Í∏∞Î≥¥Îã§ ÏûëÏúºÎ©¥ ÎÇ®ÏùÄ ÌïôÏÉù ÏàòÎ•º Í∑∏Î£π ÌÅ¨Í∏∞Î°ú ÏÑ§Ï†ï

            # Í≤∞Ï†ïÎêú Í∑∏Î£π ÌÅ¨Í∏∞ÎßåÌÅº ÌïôÏÉùÏùÑ Í∑∏Î£πÏóê Ìï†Îãπ
            group_members = all_students[:group_size]
            # Ìï†ÎãπÎêú ÌïôÏÉùÎì§ÏùÑ Ï†ÑÏ≤¥ ÌïôÏÉù Î™©Î°ùÏóêÏÑú Ï†úÍ±∞
            all_students = all_students[group_size:]

            # Í∑∏Î£πÎ™ÖÍ≥º Íµ¨ÏÑ±ÏõêÏùÑ Í∑∏Î£π ÎîïÏÖîÎÑàÎ¶¨Ïóê Ï∂îÍ∞Ä
            cls.groups[group_name] = group_members

            # Î™®Îì† ÌïôÏÉùÏù¥ Í∑∏Î£πÏóê Ìï†ÎãπÎêòÎ©¥ Î∞òÎ≥µÏùÑ Ï§ëÎã®
            if not all_students:
                break

    @classmethod
    def display_groups(cls):
        # Ï†ÄÏû•Îêú Î™®Îì† Í∑∏Î£πÍ≥º Í∑∏ Íµ¨ÏÑ±ÏõêÎì§ÏùÑ Ï∂úÎ†•
        for group_name, members in cls.groups.items():
            print(f"Í∑∏Î£π Ïù¥Î¶Ñ: {group_name}, Íµ¨ÏÑ±Ïõê: {', '.join(members)}")

    # Q 10. ÏóêÏÑú Ï∂îÍ∞Ä
    @classmethod
    def group2guild_score(cls, group_name, score):
        """
        ÏßÄÏ†ïÎêú Í∑∏Î£πÏóê ÏÜçÌïú Î©§Î≤ÑÏùò Í∏∏ÎìúÎ•º ÌôïÏù∏ÌïòÍ≥† Ï†êÏàòÎ•º Î≥ÄÍ≤Ω
        ÏßÄÏ†ïÎêú Í∑∏Î£πÏóê Ï†êÏàòÍ∞Ä Ï∂îÍ∞ÄÎêòÎ©¥ Ìï¥Îãπ Í∑∏Î£πÏóê ÏÜçÌïú Î©§Î≤ÑÎì§Ïùò Í∏∏ÎìúÏóêÎèÑ ÎèôÏùºÌïú Ï†êÏàòÍ∞Ä Ï∂îÍ∞Ä
        """
        # ÏßÄÏ†ïÎêú Í∑∏Î£π Ïù¥Î¶ÑÏù¥ cls.groupsÏóê Ï°¥Ïû¨ÌïòÎäîÏßÄ ÌôïÏù∏
        if group_name in cls.groups:
            # Í∑∏Î£πÏóê ÏÜçÌïú Î™®Îì† Î©§Î≤Ñ(ÌïôÏÉù)Î•º ÏàúÌöåÌïòÎ©∞ ÌôïÏù∏
            for student in cls.groups[group_name]:
                # Í∞Å ÌïôÏÉùÏù¥ ÏÜçÌïú Í∏∏ÎìúÎ•º Ï∞æÍ∏∞ ÏúÑÌï¥ Î™®Îì† Í∏∏ÎìúÎ•º ÏàúÌöåÌïòÎ©∞ ÌôïÏù∏
                for guild_name, members in Aiffel_Guild.guilds.items():
                    # Ìï¥Îãπ ÌïôÏÉùÏù¥ Í∏∏ÎìúÏùò Î©§Î≤Ñ Î™©Î°ùÏóê ÏûàÎäîÏßÄÏóê Îî∞Îùº Ï†êÏàò Ï∂îÍ∞Ä
                    if student in members:
                        # ÌïôÏÉùÏù¥ ÏÜçÌïú Í∏∏ÎìúÏóê Ï†êÏàòÎ•º Ï∂îÍ∞Ä
                        Aiffel_Guild.guild_score(guild_name, score)
                        """
                        Í∑∏Î£® 1Î™Ö = ÏÜåÏÜç Í∏∏Îìú 1Í∞ú -> ÏÜåÏÜç Í∏∏ÎìúÎ•º Ï∞æÏïòÏúºÎØÄÎ°ú Ï†êÏàòÎ•º Ï†ÅÏö©ÌïòÍ≥† Ï¢ÖÎ£å
                        """
                        break

    @classmethod
    def display_guilds(cls): # Ï†ÄÏû•Îêú Î™®Îì† Í∏∏ÎìúÏôÄ Í∑∏ Íµ¨ÏÑ±ÏõêÎì§ÏùÑ Ï∂úÎ†•ÌïòÎäî Î©îÏÑúÎìú
        for guild_name, members in cls.guilds.items():
            print(f"Í∏∏Îìú Ïù¥Î¶Ñ: {guild_name}, Íµ¨ÏÑ±Ïõê: {', '.join(members)}")

    # Q.11Ïóê Ï∂îÍ∞Ä : Ïò§Î≤ÑÎùºÏù¥Îî©
    @classmethod
    def display_info(cls):
        # Í∞Å Í∑∏Î£πÍ≥º Í∑∏Î£π Î©§Î≤ÑÎì§Ïùò Í∏∏Îìú Ïä§ÏΩîÏñ¥ Ï¥ùÌï©ÏùÑ ÌëúÏãú
        for group_name, members in cls.groups.items():
            print(f"Í∑∏Î£π: {group_name}")
            for member in members:
                # Î©§Î≤ÑÏùò Í∏∏Îìú Ï∞æÍ∏∞
                guild_name = next((g for g, m in Aiffel_Guild.guilds.items() if member in m), None)
                # Î©§Î≤ÑÏùò Í∏∏Îìú Ïä§ÏΩîÏñ¥ Ï¥ùÌï© ÌëúÏãú
                if guild_name:
                    score = Aiffel_Guild.guild_scores[guild_name]
                    print(f"  - {member} (Í∏∏Îìú: {guild_name}, Í∏∏Îìú Ïä§ÏΩîÏñ¥: {score})")


"""
Aiffel_GuildÏùò ÌÅ¥ÎûòÏä§ Ïò§Î≤ÑÎùºÎî©
"""
class Aiffel_Guild(Aiffel):
    guild_names = ["ÌûåÌäº", "Ïùë", "ÌïòÏÇ¨ÎπÑÏä§"]
    guilds = {guild_name: [] for guild_name in guild_names} # Í∏∏ÎìúÎ™ÖÏùÑ keyÎ°ú ÌïòÎäî dict ÏÉùÏÑ±
    guild_scores = {guild_name: 0 for guild_name in guild_names}  # Í∏∏ÎìúÎ≥Ñ Ï†êÏàò Ï¥àÍ∏∞Ìôî

    @classmethod
    def create_random_guilds(cls):
        # Aiffel ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Îì§(ÌïôÏÉùÎì§)Ïùò Ïù¥Î¶ÑÏùÑ Î¶¨Ïä§Ìä∏Î°ú Í∞ÄÏ†∏Ïò§Í∏∞(Ï†ÑÏ≤¥ Í∑∏Î£® Î™ÖÎã®)
        all_students = list(cls.instances.keys())
        random.shuffle(all_students)  # ÌïôÏÉùÎì§Ïùò ÏàúÏÑúÎ•º ÎûúÎç§ÏúºÎ°ú ÏÑûÍ∏∞

        # ÌïôÏÉùÎì§ÏùÑ ÎûúÎç§ÌïòÍ≤å Í∏∏ÎìúÏóê Î∞∞Ïπò
        for student in all_students:
            guild_name = random.choice(cls.guild_names)  # ÎûúÎç§Ìïú Í∏∏Îìú ÏÑ†ÌÉù
            cls.guilds[guild_name].append(student)  # Ìï¥Îãπ Í∏∏ÎìúÏóê ÌïôÏÉù Ï∂îÍ∞Ä

    # Q9.ÏóêÏÑú Ï∂îÍ∞Ä
    @classmethod
    def guild_score(cls, guild_name, score_change):
        # ÏßÄÏ†ïÎêú Í∏∏ÎìúÏùò Ï†êÏàòÎ•º Î≥ÄÍ≤Ω
        if guild_name in cls.guild_scores:
            cls.guild_scores[guild_name] += score_change

            # Í∏∏ÎìúÏóê ÏÜçÌïú Í∑∏Î£®Ïùò Ï†êÏàòÎèÑ Î≥ÄÍ≤Ω
            for student in cls.guilds[guild_name]:
                cls.instances[student].score_points += score_change

    @classmethod
    def display_guilds(cls): # Ï†ÄÏû•Îêú Î™®Îì† Í∏∏ÎìúÏôÄ Í∑∏ Íµ¨ÏÑ±ÏõêÎì§ÏùÑ Ï∂úÎ†•ÌïòÎäî Î©îÏÑúÎìú
        for guild_name, members in cls.guilds.items():
            print(f"Í∏∏Îìú Ïù¥Î¶Ñ: {guild_name}, Íµ¨ÏÑ±Ïõê: {', '.join(members)}")

    # Q.11Ïóê Ï∂îÍ∞Ä : Ïò§Î≤ÑÎùºÏù¥Îî©
    @classmethod
    def display_info(cls):
        # Í∞Å Í∏∏ÎìúÏùò Ï†ïÎ≥¥ÏôÄ Í∏∏ÎìúÎ≥Ñ Ïä§ÏΩîÏñ¥ Ï¥ùÌï©ÏùÑ ÌëúÏãú
        for guild_name, score in cls.guild_scores.items():
            print(f"Í∏∏Îìú: {guild_name}, Í∏∏Îìú Ïä§ÏΩîÏñ¥ Ï¥ùÌï©: {score}")


# Í≤∞Í≥º ÌôïÏù∏
# ÎûúÎç§ Í∏∏Îìú Î∞è Í∑∏Î£π ÏÉùÏÑ±
Aiffel_Guild.create_random_guilds()
Aiffel_Group.create_random_groups()

# Ï¥àÍ∏∞ Í∏∏Îìú Î∞è Í∑∏Î£π Ï†ïÎ≥¥ ÌëúÏãú
print("Ï¥àÍ∏∞ Í∏∏Îìú Ï†ïÎ≥¥:")
Aiffel_Guild.display_info()
print("\nÏ¥àÍ∏∞ Í∑∏Î£π Ï†ïÎ≥¥:")
Aiffel_Group.display_info()

# ÌäπÏ†ï Í∏∏ÎìúÏóê Ï†êÏàò Î∂ÄÏó¨ (ÏòàÏãúÎ°ú 'ÌûåÌäº' Í∏∏ÎìúÏóê +3 Ï†ê)
Aiffel_Guild.guild_score("ÌûåÌäº", 3)

# Ï†êÏàò Î≥ÄÍ≤Ω ÌõÑ Í∏∏Îìú Î∞è Í∑∏Î£π Ï†ïÎ≥¥ ÌëúÏãú
print("\nÏ†êÏàò Î≥ÄÍ≤Ω ÌõÑ Í∏∏Îìú Ï†ïÎ≥¥:")
Aiffel_Guild.display_info()
print("\nÏ†êÏàò Î≥ÄÍ≤Ω ÌõÑ Í∑∏Î£π Ï†ïÎ≥¥:")
Aiffel_Group.display_info()
```
