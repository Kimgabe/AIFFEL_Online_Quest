### Q2. 클래스 변수 생성 : Account 클래스로부터 생성된 객체의 개수를 저장하기

- Account 클래스로부터 생성된 객체의 총 개수를 추적해야 합니다.

### Q3. 클래스 변수 출력 메서드 만들기 : Account 클래스에 생성된 계좌의 계수를 출력하는 get_account_num() 메서드 추가하기

- Q2. 에서는 클래스에 선언된 변수 자체를 그대로 출력했는데, 이를 출력하는 메서드를 직접 생성하여 출력할 수 있도록 해야 합니다.
- 이를 위해 get_account_num이라는 메서드를 추가해 account_count값을 반환하도록 합니다.
    - 처음에 일반 메서드로 구현을 했는데, 값을 출력하기 위해서는 개별 인스턴스에 접근해서 출력을 해야 해서, 출력 코드가 복잡해집니다.
    - 구글링을 해보니 클래스 메서드로 구현을 하면, 인스턴스에 접근하지 않고 바로 클래스가 보유한 정보를 출력할 수 있다고 합니다.
    - 이를 반영해서 get_account_num를 클래스 메서드로 구현하였습니다.
    - get_account_num 은 데코레이터로서 사용되는 클래스 메서드가 됩니다.(인스턴스 호출없이도 사용 가능)

```python
import random # 계좌번호 랜덤 생성용

class Account:
    # 계좌 계수 출력위한 변수
    account_count = 0 # 변수 초기화

    # 생성자 생성
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # 계좌 최초 개설시 0원이므로
        self.bank_name = 'SC은행' # 은행 이름 고정
        self.account_number = self._generate_account_number()

        # Q2. 새 계좌 생성시 마다 계좌 개수 +1
        Account.account_count += 1


    # 랜덤 계좌번호 생성 메서드
    def _generate_account_number(self):
        """
        계좌번호 생성 규칙 : 3자리-2자리-6자리
        random함수의 randint 메서드 활용
        숫자의 범위는 자리수에 따라설정
        구조를 직관적으로 보기 위해 fstring사용 (일반 print문이면 ""와 ,를 섞어야 해서 가독성 떨어짐)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # Q3. 계좌 개수 출력 메서드 추가 -> 추가된 부분
    @classmethod
    def get_account_num(cls):
        return cls.account_count




# 테스트
# 여러명의 고객 객체 생성
test_accounts = [Account("고객1"), Account("고객2"), Account("고객3")]

# 랜덤 계좌정보 출력
account_numbers = [account.account_number for account in test_accounts]
#display(account_numbers)

# 인스턴스의 속성 정보 각각 출력
test_accounts_info = [(account.account_number, account.__dict__) for account in test_accounts]
#display(test_accounts_info)

account_cnt = Account.get_account_num()
print(f" 현재 Account 클래스에 저장된 계좌정보는 총 {account_cnt} 개 입니다.")
```

### Q4. 입금 메서드(deposit) 만들기 : Account 클래스에 입금을 위한 deposit 메서드 추가. 입금액의 최소단위는 1만원입니다.

- 일반 메서드로 입금액을 입력받고, 금액이 1만원 이상인지 검증하여 그 결과를 잔액에 추가해야 합니다.
- 즉, 총 3개의 기능이 구현되어야 합니다.
    - 입금액 입력받는 기능
    - 입금액 검증 하는 기능
    - 입금된 금액을 잔고에 업데이트 하는 기능
- 부가적으로 2가지 출력을 입력하겠습니다.
    - 입금전 잔액 출력
    - 입금 성공 및 실패 여부에 따른 별도의 메시지를 출력

```python
import random # 계좌번호 랜덤 생성용

class Account:
    # 계좌 계수 출력위한 변수
    account_count = 0 # 변수 초기화

    # 생성자 생성
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # 계좌 최초 개설시 0원이므로
        self.bank_name = 'SC은행' # 은행 이름 고정
        self.account_number = self._generate_account_number()

        # 새 계좌 생성시 마다 계좌 개수 +1
        Account.account_count += 1


    # 랜덤 계좌번호 생성 메서드
    def _generate_account_number(self):
        """
        계좌번호 생성 규칙 : 3자리-2자리-6자리
        random함수의 randint 메서드 활용
        숫자의 범위는 자리수에 따라설정
        구조를 직관적으로 보기 위해 fstring사용 (일반 print문이면 ""와 ,를 섞어야 해서 가독성 떨어짐)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # 계좌 개수 출력 메서드 추가
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. 입금 메서드 추가
    def deposit(self, amount):
        # 입금전 현재 잔액 정보 출력
        print("입금 전 잔액:", "{:,}원".format(self.balance)) # 천단위 쉼표 기능 구현(Q6 대비)

        # 최소 입금액 기준 충족 여부 판단
        if amount >= 1:
            self.balance += amount # 잔고에 입금액 추가
            # 결과 출력
            print(f"입금 성공! - 현재 잔액은 {self.balance:,}원 입니다.") # 천단위 쉼표 기능 구현(Q6 대비)
        # 최소 입금액 미충족시 문구 출력
        else:
            print(f"입금 실패! - 입금 시도액: {amount}원. 최소 입금액은 1원입니다. 재입금을 시도해주세요!")



# 테스트
customer1 = Account("김승순") # 신규고객 생성
customer2 = Account("김아이펠") # 신규고객 생성

# 고객 1 20,000원 입금 시도
customer1.deposit(20000)

# 고객 2 0원 입금 시도
customer2.deposit(0)
```

- 처음 의도한대로 입금전 잔액을 출력하였으며, 입금액에 따라 성공 및 실패에 따른 각각의 메시지를 출력하도록 했습니다.
- 금액에 천단위로 쉼표도 Q6에 구현하여야 하는 부분인데 지속적으로 사용하기 위해 미리 입력하였습니다.

- 기능 구현을 위해 필요한 사항은 다음과 같습니다.
    - 입금 횟수 추적 : deposit이 실행될때마다 입금 횟수를 추적하는 인스턴스 변수가 추가되어야 합니다.
    - 이자 지급을 위한 로직 추가 : 입금횟수 5회마다 잔액을 조회하고, 그 잔액(5회차 입금액 포함)의 1%를 추가로 입금해야 합니다.
    - 지급한 이자를 잔고에 업데이트 해야 합니다.
- 각 기능에서 필요한 print 문도 적절하게 포함하여 추가하겠습니다.
    - 이자 지급 조건 충족 여부
    - 이자 금액(소수점2자리)
- 부가적으로 이자로 인해 소수점이 늘어나므로, 이후의 금액 출력관련 부분은 소수점 2자리까지만 출력하는것으로 일괄 수정했습니다.

```python
import random # 계좌번호 랜덤 생성용

class Account:
    # 계좌 계수 출력위한 변수
    account_count = 0 # 변수 초기화

    # 생성자 생성
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # 계좌 최초 개설시 0원이므로
        self.bank_name = 'SC은행' # 은행 이름 고정
        self.account_number = self._generate_account_number()

        # 새 계좌 생성시 마다 계좌 개수 +1
        Account.account_count += 1

        # Q7에서 추가된 코드
        self.deposit_count = 0 # 입금 횟수 추적 위한 변수


    # 랜덤 계좌번호 생성 메서드
    def _generate_account_number(self):
        """
        계좌번호 생성 규칙 : 3자리-2자리-6자리
        random함수의 randint 메서드 활용
        숫자의 범위는 자리수에 따라설정
        구조를 직관적으로 보기 위해 fstring사용 (일반 print문이면 ""와 ,를 섞어야 해서 가독성 떨어짐)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # 계좌 개수 출력 메서드 추가
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. 입금 메서드 추가
    # Q7. 이자지급 메서드를 deposit의 하위 메서드로 추가
    def deposit(self, amount):
        # 입금전 현재 잔액 정보 출력
        print("입금 전 잔액:", "{:,}원".format(self.balance)) # 천단위 쉼표 기능 구현(Q6 대비)

        # 최소 입금액 기준 충족 여부 판단
        if amount >= 1:
            self.balance += amount # 잔고에 입금액 추가
            self.deposit_count += 1 # 입금횟수 증가
            # 결과 출력
            print(f"입금 성공! - 현재 잔액은 {self.balance:,.2f}원 입니다.") # 천단위 쉼표 기능 구현(Q6 대비)


            # Q7. 입금 5회당 이자 지급 기능 추가하는 코드
            if self.deposit_count % 5 == 0: # 5회당 이므로 5로 나누어 나머지가 0이되어야 함
                interest = self.balance * 0.01 # 잔고(balance)를 불러오고 해당 금액의 1% 이자로 계산
                self.balance += interest # 이자 지급 및 반영
                print(f"누적 입금 시도가 {self.deposit_count}회입니다! 내부정책에 따라 현재 잔고의 1%를 이자로 지급합니다.")
                print(f"추가 이자 지급: {interest:,.2f}원 - 새로운 잔액: {self.balance:,.2f}원")

        # 최소 입금액 미충족시 문구 출력
        else:
            print(f"입금 실패! - 입금 시도액: {amount}원. 최소 입금액은 1원입니다. 재입금을 시도해주세요!")


    # Q5. 출금 메서드 추가
    def withdraw(self, amount):
        # 현재 잔고 출력
        print("현재 잔고 :", "{:,}원".format(self.balance))

        # 출고 요청 금액에 대한 출금 여부 검증
        if amount <= self.balance:
            self.balance -= amount # 출금요청금액을 잔고에 반영(차감)
            print(f"출금 성공! - 현재 잔액은 {self.balance:,.2f}원 입니다.")
        else:
            print(f"출금 실패! - 출금 시도액: {amount}원. 잔액이 부족합니다.")


    # Q6. 정보출력 메서드 구현하기
    def display_info(self):
        # 각각의 정보를 출력하는 print문 작성
        print(f"은행 이름: {self.bank_name}")
        print(f"예금주: {self.account_name}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔액: {self.balance:,.2f}원")



# 테스트
customer1 = Account("김승순") # 신규고객 생성

# 5회 입금 시도
for _ in range(5):
    customer1.deposit(10000) # 매번 만원씩 입금

# 계좌 정보 출력
customer1.display_info()
```

## Q10. 입출금 내역 코드 업데이트 : 입금내역을 출력하는 deposit_history 메서드, 출금내역을 출력하는 withdraw_history 메서드를 추가 하세요.

- 입출금 내역을 기록하였다가 출력하는 메서드를 각각 만들어야 합니다.
- 이를 위해 입/출금시 발생하는 각 거래 내역을 저장할 변수를 생성자에 추가해야 합니다.
- 입/출금시마다 발생하는 정보를
- 또한 해당 정보를 불러와서

```python
import random # 계좌번호 랜덤 생성용

class Account:
    # 계좌 계수 출력위한 변수
    account_count = 0 # 변수 초기화

    # 생성자 생성
    def __init__(self, account_name):
        self.account_name = account_name
        self.balance = 0 # 계좌 최초 개설시 0원이므로
        self.bank_name = 'SC은행' # 은행 이름 고정
        self.account_number = self._generate_account_number()

        # 새 계좌 생성시 마다 계좌 개수 +1
        Account.account_count += 1

        # Q7에서 추가된 코드
        self.deposit_count = 0 # 입금 횟수 추적 위한 변수

        # Q10. 에서 추가된 코드
        self.deposit_history_list = []  # 입금 내역 저장을 위한 리스트
        self.withdraw_history_list = []  # 출금 내역 저장을 위한 리스트


    # 랜덤 계좌번호 생성 메서드
    def _generate_account_number(self):
        """
        계좌번호 생성 규칙 : 3자리-2자리-6자리
        random함수의 randint 메서드 활용
        숫자의 범위는 자리수에 따라설정
        구조를 직관적으로 보기 위해 fstring사용 (일반 print문이면 ""와 ,를 섞어야 해서 가독성 떨어짐)
        """
        return f"{random.randint(100,999)}-{random.randint(10,99)}-{random.randint(100000,999999)}"

    # 계좌 개수 출력 메서드 추가
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4. 입금 메서드 추가
    # Q7. 이자지급 메서드를 deposit의 하위 메서드로 추가
    def deposit(self, amount):
        # 입금전 현재 잔액 정보 출력
        print("입금 전 잔액:", "{:,}원".format(self.balance)) # 천단위 쉼표 기능 구현(Q6 대비)

        # 최소 입금액 기준 충족 여부 판단
        if amount >= 1:
            self.balance += amount # 잔고에 입금액 추가
            self.deposit_count += 1 # 입금횟수 증가
            self.deposit_history_list.append(amount)  # 입금 내역에 추가
            # 결과 출력
            print(f"입금 성공! - 현재 잔액은 {self.balance:,.2f}원 입니다.") # 천단위 쉼표 기능 구현(Q6 대비)


            # Q7. 입금 5회당 이자 지급 기능 추가하는 코드
            if self.deposit_count % 5 == 0: # 5회당 이므로 5로 나누어 나머지가 0이되어야 함
                interest = self.balance * 0.01 # 잔고(balance)를 불러오고 해당 금액의 1% 이자로 계산
                self.balance += interest # 이자 지급 및 반영
                print(f"누적 입금 시도가 {self.deposit_count}회입니다! 내부정책에 따라 현재 잔고의 1%를 이자로 지급합니다.")
                print(f"추가 이자 지급: {interest:,.2f}원 - 새로운 잔액: {self.balance:,.2f}원")

        # 최소 입금액 미충족시 문구 출력
        else:
            print(f"입금 실패! - 입금 시도액: {amount}원. 최소 입금액은 1원입니다. 재입금을 시도해주세요!")


    # Q5. 출금 메서드 추가
    def withdraw(self, amount):
        # 현재 잔고 출력
        print("현재 잔고 :", "{:,}원".format(self.balance))

        # 출고 요청 금액에 대한 출금 여부 검증
        if amount <= self.balance:
            self.balance -= amount # 출금요청금액을 잔고에 반영(차감)
            # Q10. 출금 내역 history로 저장
            self.withdraw_history_list.append(amount)
            print(f"출금 성공! - 현재 잔액은 {self.balance:,.2f}원 입니다.")
        else:
            print(f"출금 실패! - 출금 시도액: {amount}원. 잔액이 부족합니다.")


    # Q6. 정보출력 메서드 구현하기
    def display_info(self):
        # 각각의 정보를 출력하는 print문 작성
        print(f"은행 이름: {self.bank_name}")
        print(f"예금주: {self.account_name}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔액: {self.balance:,.2f}원")


    # Q10. 입금 내역 메서드 추가
    def deposit_history(self):
        for deposit in self.deposit_history_list: # 생성자에 있는 deposit_history_list를 iteration하며 정보 조회
            print(f" - {deposit:,.2f}원 입금") # 각 입금 기록을 출력
        if not self.deposit_history_list: # 아직 기록이 없는 경우
            print("입금 내역이 없습니다.")

    # Q10. 출금 내역 메서드 추가
    def withdraw_history(self):
        for withdraw in self.withdraw_history_list: # 생성자에 있는 withdraw_history_list를 iteration하며 정보 조회
            print(f" - {withdraw:,.2f}원 출력") # 각 출금 기록을 출력
        if not self.withdraw_history_list: # 아직 기록이 없는 경우
            print("출금 내역이 없습니다.")


# 테스트
customer1 = Account("김승순")
customer1.deposit(10000)  # 입금
customer1.withdraw(5000)  # 출금
customer1.deposit(20000)  # 입금

# 입금 및 출금 내역 출력
customer1.deposit_history()
customer1.withdraw_history()
```

- 처음에 의도 했던 대로 각각의 입출력 내용이 리스트에 저장되어있다가 출력되고 있음을 확인할 수 있습니다.

### **Q1. 코어 명단(딕셔너리)를 입력받아 학번을 부여하는 초기화 메서드를 작성해 보세요**

- 일단 Aiffel클래스와 인스턴스를 생성해야 합니다.
- 이후 사전에 지정한 코어명단 딕셔너리를 입력받아 정해진 규칙에 따라 학번을 생성해야 합니다.
- 학번의 경우 가나다 규칙을 제외하고는 모두 같은 값이므로 일괄입력받아 적용하도록 하겠습니다.
- 그 외 이름의 경우 이미 적용이 되어있긴하나, 차후 추가 사용 가능성을 고려해 가다다순 정렬을 적용하도록 하겠습니다.
- 이 로직이 적용된 학번 생성 기능을 generate_student_id 라는 메서드로 추가하겠습니다.

```python
class Aiffel:
    # 생성된 학번의 총 개수
    total_id_count = 0  # 초기화

    def __init__(self, name, admission_year="2023", term=29, course_type="코어"):
        self.name = name  # 학생의 이름
        self.admission_year = admission_year  # 입학 연도
        self.term = term  # 기수
        self.course_type = course_type  # 과정 종류
        self.student_id = self.generate_student_id()  # 학번 생성 메서드 호출
        Aiffel.total_id_count += 1  # 학번이 생성될 때마다 카운트 증가

    def generate_student_id(self):
        year = self.admission_year[-2:]  # 입학 연도의 마지막 두 자리
        course_type_num = '2' if self.course_type == "코어" else '1'  # 과정 종류에 따른 숫자 할당
        student_number = str(Aiffel.total_id_count + 1).zfill(2)  # 학번의 순서를 나타내는 두 자리 숫자
        student_id = f"{year}{str(self.term).zfill(2)}{course_type_num}{student_number}"  # 학번 조합
        return student_id  # 생성된 학번 반환

# 코어 명단 dict
core_members =  {'강영현': 0, '김경민': 0, '김민기': 0, '김승순': 0, '김양희': 0, '김영신': 0,
                 '오우진': 0, '윤혜윤': 0, '이상원': 0, '이선재': 0, '이슬': 0, '이승환': 0, '전민규': 0,
                 '조보겸': 0, '조수민': 0, '최강훈': 0, '한현종': 0, '홍채림': 0}

# 코어 명단을 바탕으로 Aiffel 인스턴스 생성
members_instances = {} # 코어 명단 dict를 바탕으로 Aiffel 클래스의 인스턴스를 생성용 dict

# 각 학생의 이름을 사용하여 Aiffel 클래스의 인스턴스를 생성
# Aiffel 클래스의 생성자에 정의된 기본값(연도, 기수, 과정 종류) 적용
for name in core_members.keys():
    members_instances[name] = Aiffel(name)

# 생성된 인스턴스(그루)의 학번 확인
for name, instance in members_instances.items():
    print(f"이름: {name}, 학번: {instance.student_id}")
```

## **Q2. Aiffel 클래스로부터 생성된 학번의 개수를 출력하는 get_id_num() 메서드를 추가하세요.**

- 생성자 메서드에서 학번 생성시마다 위에 추가한 total_id_count 변수의 값을 증가하도록 합니다.

```python
import pandas as pd # csv 형태의 데이터 출력용

class Aiffel:
    total_id_count = 0  # 생성된 학번의 총 개수
    instances = {}  # 생성된 Aiffel 인스턴스 목록을 클래스 레벨에서 관리
    members_instances = {} # 코어 명단 dict를 바탕으로 Aiffel 클래스의 인스턴스를 생성용 dict

    def __init__(self, name, admission_year="2023", term=29, course_type="코어"):
        self.name = name  # 학생의 이름
        self.admission_year = admission_year  # 입학 연도
        self.term = term  # 기수
        self.course_type = course_type  # 과정 종류
        self.student_id = self.generate_student_id()  # 학번 생성 메서드 호출
        self.score_points = 0  # 개별 그루의 퀘스트 점수
        self.penalty_points = 0  # 개별 그루의 퀘스트 penalty
        Aiffel.instances[name] = self  # 생성된 인스턴스를 이름을 key로 딕셔너리에 추가
        Aiffel.total_id_count += 1  # 학번 생성시마다 총 학번 개수 증가

    def generate_student_id(self):
        year = self.admission_year[-2:]  # 입학 연도의 마지막 두 자리
        course_type_num = '2' if self.course_type == "코어" else '1'  # 과정 종류에 따른 숫자 할당
        student_number = str(Aiffel.total_id_count + 1).zfill(2)  # 학번의 순서를 나타내는 두 자리 숫자
        student_id = f"{year}{str(self.term).zfill(2)}{course_type_num}{student_number}"  # 학번 조합
        return student_id  # 생성된 학번 반환

    @classmethod
    def get_id_num(cls):
        return cls.total_id_count

    # Q3. 에서 추가
    def score(self, points):
        if points in [1, 2, 3]:
            self.score_points += points  # 조건에 해당할 경우 점수 추가
        else:
            print("점수는 1점, 2점, 3점 중 1개만 입력 가능합니다.")

    # Q4. 에서 추가
    def penalty(self, points):
        self.penalty_points -= points  # 입력받은 점수를 차감

    # Q5. 에서 추가
    @classmethod
    def display_info(cls):
        data = {
            "이름": [],
            "학번": [],
            "과정 종류": [],
            "퀘스트 점수 총점": [],
            "퀘스트 입력 총점": [],
            "퀘스트 penalty 총합": []
        }

        # 클래스 내부에서 instances 변수에 접근
        for name in cls.instances.keys():
            student = cls.instances[name]
            data["이름"].append(student.name)
            data["학번"].append(student.student_id)
            data["과정 종류"].append(student.course_type)
            data["퀘스트 점수 총점"].append(student.score_points + student.penalty_points)
            data["퀘스트 입력 총점"].append(student.score_points)
            data["퀘스트 penalty 총합"].append(student.penalty_points)

        df = pd.DataFrame(data)
        display(df)

    # Q6. 중도포기자 관리를 위한 메서드 추가 - 명단 삭제
    @classmethod
    def remove_member(cls, name):
        if name in cls.instances:
            del cls.instances[name]  # 해당 그루 정보 삭제
            print(f"{name} 님의 정보가 삭제되었습니다.")
        else:
            print(f"{name} 님은 명단에 존재하지 않습니다.")


# 클래스내에 그루 정보 생성 및 저장
# 코어 명단 dict
core_members =  {'강영현': 0, '김경민': 0, '김민기': 0, '김승순': 0, '김양희': 0, '김영신': 0,
                 '오우진': 0, '윤혜윤': 0, '이상원': 0, '이선재': 0, '이슬': 0, '이승환': 0, '전민규': 0,
                 '조보겸': 0, '조수민': 0, '최강훈': 0, '한현종': 0, '홍채림': 0}

# 코어 명단을 바탕으로 Aiffel 인스턴스 생성

# 각 학생의 이름을 사용하여 Aiffel 클래스의 인스턴스를 생성
# Aiffel 클래스의 생성자에 정의된 기본값(연도, 기수, 과정 종류) 적용
for name in core_members.keys():
    members_instances[name] = Aiffel(name)

# 학생 정보 조회
Aiffel.display_info()

# remove_member테스트
# 김경민과 김영신을 삭제
Aiffel.remove_member("김경민")
Aiffel.remove_member("김영신")

# 갱신된 명단 출력
Aiffel.display_info()

# 다시한번 삭제 시도
Aiffel.remove_member("김영신")
```

## **Q7. Aiffel 클래스를 상속하여, 그루 명단을 바탕으로 2인, 3인, 4인으로 랜덤 그룹핑을 해주는 메서드가 추가된 Aiffel_Group 클래스를 만들어 보세요. 그룹이름은 총 8개입니다. (상생, 신뢰, 열정, 이끔, 성장, 가치, 공유, 확산)**

- Aiffel 클래스를 상속하며 모든 정보를 다 승계 받아야 합니다.
- 랜덤 그루핑을 위한 메서드가 추가되어야 합니다.
    - create_random_groupe라는 메서드를 추가하겠습니다.
    - 해당 메서드는 전체 그루 인원정보에서 2인, 3인, 4인의 조합을 랜덤으로 하여 8개의 그룹에 랜덤으로 배치합니다.
    - 즉, 그룹에 배치할 인원을 뽑는 것도 랜덤, 어느 그룹에 배치할지도 랜덤이 되어야 합니다.
    - 다만, 배치될 그룹은 8개가 모두 처음부터 생성되어 있어야 합니다.
- 생성된 그룹 정보는 클래스의 변수로 저장해서 관리합니다.

```python
import random

class Aiffel_Group(Aiffel):
    # 그룹 이름을 미리 정의
    group_names = ["상생", "신뢰", "열정", "이끔", "성장", "가치", "공유", "확산"]
    # 생성된 그룹을 저장할 딕셔너리를 초기화
    groups = {}

    @classmethod
    def create_random_groups(cls):
        # Aiffel 클래스의 인스턴스들(학생들)의 이름을 리스트로 가져오기
        all_students = list(cls.instances.keys())
        # 학생들의 순서를 랜덤으로 섞기
        random.shuffle(all_students)


        # 먼저 8개의 그룹을 생성
        for group_name in cls.group_names:
            cls.groups[group_name] = []



        # 모든 학생을 그룹에 랜덤으로 배치
        while all_students:
            for group_name in cls.group_names:
                if not all_students:
                    break
                # 각 그룹에 2명, 3명 또는 4명의 학생을 랜덤으로 배치
                group_size = random.choice([2, 3, 4])
                # 남은 학생 수가 결정된 그룹 크기보다 작으면, 남은 학생 수를 그룹 크기로 설정
                group_size = min(group_size, len(all_students))

                # 학생들을 그룹에 할당
                cls.groups[group_name].extend(all_students[:group_size])
                all_students = all_students[group_size:]

    @classmethod
    def display_groups(cls):
        # 저장된 모든 그룹과 그 구성원들을 출력
        for group_name, members in cls.groups.items():
            print(f"그룹 이름: {group_name}, 구성원: {', '.join(members)}")

# Aiffel_Group 클래스의 랜덤 그룹 생성 및 출력 메서드를 호출
Aiffel_Group.create_random_groups()
Aiffel_Group.display_groups()
```

```python
class Aiffel_Group(Aiffel):
    # 그룹 이름을 미리 정의
    group_names = ["상생", "신뢰", "열정", "이끔", "성장", "가치", "공유", "확산"]
    # 생성된 그룹을 저장할 딕셔너리를 초기화
    groups = {}

    @classmethod
    def create_random_groups(cls):
        # Aiffel 클래스의 인스턴스들(학생들)의 이름을 리스트로 가져오기
        all_students = list(cls.instances.keys())
        # 학생들의 순서를 랜덤으로 섞기
        random.shuffle(all_students)

        # 정의된 그룹 이름만큼 반복하여 각 그룹에 학생들을 할당
        for group_name in cls.group_names:
            # 그룹의 구성원 수를 2, 3, 4 중 랜덤으로 결정
            group_size = random.choice([2, 3, 4])
            if group_size > len(all_students):
                group_size = len(all_students) # 만약 남은 학생 수가 결정된 그룹 크기보다 작으면 남은 학생 수를 그룹 크기로 설정

            # 결정된 그룹 크기만큼 학생을 그룹에 할당
            group_members = all_students[:group_size]
            # 할당된 학생들을 전체 학생 목록에서 제거
            all_students = all_students[group_size:]

            # 그룹명과 구성원을 그룹 딕셔너리에 추가
            cls.groups[group_name] = group_members

            # 모든 학생이 그룹에 할당되면 반복을 중단
            if not all_students:
                break

    @classmethod
    def display_groups(cls):
        # 저장된 모든 그룹과 그 구성원들을 출력
        for group_name, members in cls.groups.items():
            print(f"그룹 이름: {group_name}, 구성원: {', '.join(members)}")

    # Q 10. 에서 추가
    @classmethod
    def group2guild_score(cls, group_name, score):
        """
        지정된 그룹에 속한 멤버의 길드를 확인하고 점수를 변경
        지정된 그룹에 점수가 추가되면 해당 그룹에 속한 멤버들의 길드에도 동일한 점수가 추가
        """
        # 지정된 그룹 이름이 cls.groups에 존재하는지 확인
        if group_name in cls.groups:
            # 그룹에 속한 모든 멤버(학생)를 순회하며 확인
            for student in cls.groups[group_name]:
                # 각 학생이 속한 길드를 찾기 위해 모든 길드를 순회하며 확인
                for guild_name, members in Aiffel_Guild.guilds.items():
                    # 해당 학생이 길드의 멤버 목록에 있는지에 따라 점수 추가
                    if student in members:
                        # 학생이 속한 길드에 점수를 추가
                        Aiffel_Guild.guild_score(guild_name, score)
                        """
                        그루 1명 = 소속 길드 1개 -> 소속 길드를 찾았으므로 점수를 적용하고 종료
                        """
                        break

    @classmethod
    def display_guilds(cls): # 저장된 모든 길드와 그 구성원들을 출력하는 메서드
        for guild_name, members in cls.guilds.items():
            print(f"길드 이름: {guild_name}, 구성원: {', '.join(members)}")



# 결과 확인

# 길드 생성 및 초기 길드 정보 표시
Aiffel_Guild.create_random_guilds()
Aiffel_Guild.display_guilds()

# 랜덤 그룹 생성 및 초기 그룹 정보 표시
Aiffel_Group.create_random_groups()
Aiffel_Group.display_groups()

# 초기 그루 점수 정보 표시
Aiffel.display_info()

# '상상' 그룹에 +2 점수 부여
Aiffel_Group.group2guild_score("상생", 2)

# 점수 부여 후 길드 및 그루 점수 정보 재표시
Aiffel.display_info()
```

## **11. Aiffel_Group 클래스와 Aiffel_Guild클래스가 Aiffel 클래스의 display_info() 메서드를 아래와 같이 오버라이딩 하도록 합니다.**

- 기존 명단에 길드별 스코어 총합이 보이도록 명단을 출력하도록 display_info()를 오버라이딩 합니다.
---

- 문제 풀이를 위해 아래의 작업을 진행해야 합니다.
    - Aiffel_Group 클래스 오버라이딩: display_info() 메서드를 오버라이딩하여 각 그룹의 정보와 그룹 멤버들의 길드 스코어 총합을 표시
    - Aiffel_Guild 클래스 오버라이딩: display_info() 메서드를 오버라이딩하여 각 길드의 정보와 길드별 스코어 총합을 표시

```python
"""
Aiffel_Group 의 정보 오버라이딩
"""
class Aiffel_Group(Aiffel):
    # 그룹 이름을 미리 정의
    group_names = ["상생", "신뢰", "열정", "이끔", "성장", "가치", "공유", "확산"]
    # 생성된 그룹을 저장할 딕셔너리를 초기화
    groups = {}

    @classmethod
    def create_random_groups(cls):
        # Aiffel 클래스의 인스턴스들(학생들)의 이름을 리스트로 가져오기
        all_students = list(cls.instances.keys())
        # 학생들의 순서를 랜덤으로 섞기
        random.shuffle(all_students)

        # 정의된 그룹 이름만큼 반복하여 각 그룹에 학생들을 할당
        for group_name in cls.group_names:
            # 그룹의 구성원 수를 2, 3, 4 중 랜덤으로 결정
            group_size = random.choice([2, 3, 4])
            if group_size > len(all_students):
                group_size = len(all_students) # 만약 남은 학생 수가 결정된 그룹 크기보다 작으면 남은 학생 수를 그룹 크기로 설정

            # 결정된 그룹 크기만큼 학생을 그룹에 할당
            group_members = all_students[:group_size]
            # 할당된 학생들을 전체 학생 목록에서 제거
            all_students = all_students[group_size:]

            # 그룹명과 구성원을 그룹 딕셔너리에 추가
            cls.groups[group_name] = group_members

            # 모든 학생이 그룹에 할당되면 반복을 중단
            if not all_students:
                break

    @classmethod
    def display_groups(cls):
        # 저장된 모든 그룹과 그 구성원들을 출력
        for group_name, members in cls.groups.items():
            print(f"그룹 이름: {group_name}, 구성원: {', '.join(members)}")

    # Q 10. 에서 추가
    @classmethod
    def group2guild_score(cls, group_name, score):
        """
        지정된 그룹에 속한 멤버의 길드를 확인하고 점수를 변경
        지정된 그룹에 점수가 추가되면 해당 그룹에 속한 멤버들의 길드에도 동일한 점수가 추가
        """
        # 지정된 그룹 이름이 cls.groups에 존재하는지 확인
        if group_name in cls.groups:
            # 그룹에 속한 모든 멤버(학생)를 순회하며 확인
            for student in cls.groups[group_name]:
                # 각 학생이 속한 길드를 찾기 위해 모든 길드를 순회하며 확인
                for guild_name, members in Aiffel_Guild.guilds.items():
                    # 해당 학생이 길드의 멤버 목록에 있는지에 따라 점수 추가
                    if student in members:
                        # 학생이 속한 길드에 점수를 추가
                        Aiffel_Guild.guild_score(guild_name, score)
                        """
                        그루 1명 = 소속 길드 1개 -> 소속 길드를 찾았으므로 점수를 적용하고 종료
                        """
                        break

    @classmethod
    def display_guilds(cls): # 저장된 모든 길드와 그 구성원들을 출력하는 메서드
        for guild_name, members in cls.guilds.items():
            print(f"길드 이름: {guild_name}, 구성원: {', '.join(members)}")

    # Q.11에 추가 : 오버라이딩
    @classmethod
    def display_info(cls):
        # 각 그룹과 그룹 멤버들의 길드 스코어 총합을 표시
        for group_name, members in cls.groups.items():
            print(f"그룹: {group_name}")
            for member in members:
                # 멤버의 길드 찾기
                guild_name = next((g for g, m in Aiffel_Guild.guilds.items() if member in m), None)
                # 멤버의 길드 스코어 총합 표시
                if guild_name:
                    score = Aiffel_Guild.guild_scores[guild_name]
                    print(f"  - {member} (길드: {guild_name}, 길드 스코어: {score})")


"""
Aiffel_Guild의 클래스 오버라딩
"""
class Aiffel_Guild(Aiffel):
    guild_names = ["힌튼", "응", "하사비스"]
    guilds = {guild_name: [] for guild_name in guild_names} # 길드명을 key로 하는 dict 생성
    guild_scores = {guild_name: 0 for guild_name in guild_names}  # 길드별 점수 초기화

    @classmethod
    def create_random_guilds(cls):
        # Aiffel 클래스의 인스턴스들(학생들)의 이름을 리스트로 가져오기(전체 그루 명단)
        all_students = list(cls.instances.keys())
        random.shuffle(all_students)  # 학생들의 순서를 랜덤으로 섞기

        # 학생들을 랜덤하게 길드에 배치
        for student in all_students:
            guild_name = random.choice(cls.guild_names)  # 랜덤한 길드 선택
            cls.guilds[guild_name].append(student)  # 해당 길드에 학생 추가

    # Q9.에서 추가
    @classmethod
    def guild_score(cls, guild_name, score_change):
        # 지정된 길드의 점수를 변경
        if guild_name in cls.guild_scores:
            cls.guild_scores[guild_name] += score_change

            # 길드에 속한 그루의 점수도 변경
            for student in cls.guilds[guild_name]:
                cls.instances[student].score_points += score_change

    @classmethod
    def display_guilds(cls): # 저장된 모든 길드와 그 구성원들을 출력하는 메서드
        for guild_name, members in cls.guilds.items():
            print(f"길드 이름: {guild_name}, 구성원: {', '.join(members)}")

    # Q.11에 추가 : 오버라이딩
    @classmethod
    def display_info(cls):
        # 각 길드의 정보와 길드별 스코어 총합을 표시
        for guild_name, score in cls.guild_scores.items():
            print(f"길드: {guild_name}, 길드 스코어 총합: {score}")


# 결과 확인
# 랜덤 길드 및 그룹 생성
Aiffel_Guild.create_random_guilds()
Aiffel_Group.create_random_groups()

# 초기 길드 및 그룹 정보 표시
print("초기 길드 정보:")
Aiffel_Guild.display_info()
print("\n초기 그룹 정보:")
Aiffel_Group.display_info()

# 특정 길드에 점수 부여 (예시로 '힌튼' 길드에 +3 점)
Aiffel_Guild.guild_score("힌튼", 3)

# 점수 변경 후 길드 및 그룹 정보 표시
print("\n점수 변경 후 길드 정보:")
Aiffel_Guild.display_info()
print("\n점수 변경 후 그룹 정보:")
Aiffel_Group.display_info()
```
