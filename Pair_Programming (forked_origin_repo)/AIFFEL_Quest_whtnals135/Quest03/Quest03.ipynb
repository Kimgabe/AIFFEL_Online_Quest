```python
#데이터 불러오기
txt_file = open("/content/drive/MyDrive/Colab Notebooks/Avengers.txt")

import re


for line in txt_file:
    clean_line = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "",line) #특수기호 제거
    txt_data = clean_line.lower() #모든 문자를 소문자로 변환
    SplitTxt = txt_data.split()  #N-gram 사용전 단어별로 나누기

    #N-gram 만들기
    for i in range(len(SplitTxt) -1 ):
      a= list((SplitTxt[i], SplitTxt[i + 1]))
      print(a)

#빈도 수가 높은 단어와 그 단어의 갯수를 출력하기 위해 시도했었다,,,,
#from collections import Counter
#Counter(list((SplitTxt[i], SplitTxt[i + 1]))).most.common()



txt_file.close()


# 회고
# 처음에 파일자체를 불러오는 것을 몰라서 open함수에서 막혔는데 구글드라이브 활용하여 연결하는 것을 알게 되었다.
# 본인의 컴퓨터로 저장 후에 open함수가 안되는 이유는 colab은 구글 클라우드 서버에 연결되어있기 때문에 구글드라이브
# 에서 불러와야 파일을 열 수 있었다.처음에 구글드라이브를 몰라서 시간을 많이 사용한 것이 아쉬웠다.
# 그 후에 for문과 lower()함수를 통해 대문자를 소문자로 바꾸어주었고 split()
# 을 사용하여 단어단위로 나누어주었고 그뒤에 인덱싱하여 리스트로 묶어 2gram을 만들었지만 시간이 없어서 그 뒤로는 하지 못해서 아쉬웠다.
```
