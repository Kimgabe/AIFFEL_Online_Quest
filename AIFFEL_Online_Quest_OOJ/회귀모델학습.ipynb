```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso

# 데이터 생성
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + 1.5 * np.random.randn(100, 1)

# 선형 회귀
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print("Linear Regression Coefficients:", lin_reg.coef_, lin_reg.intercept_)

# 릿지 회귀
ridge_reg = Ridge(alpha=1)
ridge_reg.fit(X, y)
print("Ridge Regression Coefficients:", ridge_reg.coef_, ridge_reg.intercept_)

# 라쏘 회귀
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)
print("Lasso Regression Coefficients:", lasso_reg.coef_, lasso_reg.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

# 랜덤한 데이터 생성
np.random.seed(42)
X = 3 * np.random.rand(100, 1)
y = 2 + 4 * X + 1.5 * np.random.randn(100, 1)

# 라쏘 회귀 모델 학습
lasso_reg = Lasso(alpha=1)
lasso_reg.fit(X, y)

# 가중치 출력
print("Lasso Regression Coefficients:", lasso_reg.coef_, lasso_reg.intercept_)

# 결과 시각화
plt.scatter(X, y, label='Data points')
plt.plot(X, lasso_reg.predict(X), color='red', label='Lasso Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
