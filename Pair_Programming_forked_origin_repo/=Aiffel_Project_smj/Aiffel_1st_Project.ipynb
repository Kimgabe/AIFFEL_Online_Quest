## Project 01. 사각형의 넓이 구하기
  사각형의 넓이 구하기     
  직사각형과 평행사변형, (등변)사다리꼴의 넓이를 모두 구할 수 있는 클래스를 만들어 보자.  

<br>

- 직사각형: 가로 x 세로
- 평행사변형: 밑변 x 높이
- (등변) 사다리꼴: (밑변 + 윗변) x 높이 x 1/2

<br>

###🔶 사각형 클래스를 먼저 만들어 보자!  
   

   우선 세 가지 사각형의 넓이를 구할 수 있는 사각형(Square) 클래스를 만들자.
       
```python
class Square:
    def __init__(self):  # __init__의 매개변수의 첫번째는 self.
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))  # 입력값을 int로 묶기

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')  # Square 클래스에 rect() 메서드를 추가.

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')  # Square 클래스에 par() 메서드를 추가.  
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')  # Square 클래스에 trape() 메서드를 추가.
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')
```




` split()` 함수를 쓰면 쪼개진 개체들의 리스트가 나오는데, 리스트는 `int()` 함수 안에 넣을 수 없다. 이럴 때 쓰는 함수가 `map()` 이라는 함수다.  

<br>


`map()`와 ` split()` 함수를 이용해서 직사각형의 넓이를 구하는 메서드를 작성해보자.  

```python
def rect(self):  # 직사각형의 넓이를 구하는 메서드
    width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))  #가로, 세로를 입력받는 코드
    area = width * vertical # 넓이를 구하는 부분
    result = '직사각형의 넓이는 : ' + str(area)
    return result
```

이 식을 참고하여 앞서 언급했던 세 가지의 사각형의 넓이를 구해보자






````python
# 사각형 넓이를 구하는 클래스
class Square:
    def __init__(self):
        square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    #직사각형 메서드
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    #평행사변형 메서드
    def par(self):
        base, height = map(int,input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는: ' + str(area)
        return result

    # 사다리꼴 메서드
    def trape(self):
        base, top_side, height = map(int,input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))
        area = (base + top_side) * height  * 1/2
        result = '사다리꼴의 넓이는: ' +str(area)
        return result

a = Square()  # 클래스 객체 생성. 1, 2, 3을 각각 입력하여 확인
print(a.rect()) # 원하는 숫자를 입력하고 직사각형의 넓이 구하기
print(a.par()) # 원하는 숫자를 입력하고 평행사변형의 넓이 구하기
print(a.trape()) # 원하는 숫자를 입력하고 (등변)사다리꼴의 넓이 구하기
````

## Project 02. 키오스크 만들기  

<br>  

카페에 가서 음료를 주문할 때 우리에게 필요한 변수들은 무엇이 있을까?  
- HOT or ICE
- 주문 메뉴
- 주문 금액  
<br>  


우선 음료 메뉴와 음료 가격을 가져와보자  

```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]
```  
<br>
<br>



추가적으로 코드를 짤 때에 조건을 정해보자  
<br>   

- 메뉴와 가격 리스트를 다차원 리스트로 변경하기
- 메뉴출력 `enumerate()` 함수 활용하기
- `datetime` 모듈로 주문표에 주문 일시 추가하기
- 데코레이터로 주문표 양식 출력하기  

<br>
다차원 리스트로  self.menu = [['americano', 2000], ['latte', 3000], ['mocha', 3000], ['yuza_tea', 2500], ['green_tea', 2500], ['choco_latte', 3000]]를 만들고
메뉴 출력, 메뉴 주문, 음료 온도 체크, 추가 주문, 지불 순서대로 코드를 짜보자.  






````python
from datetime import datetime
class Kiosk:
    def __init__(self):
         self.menu = [
            ['americano', 2000],
            ['latte', 3000],
            ['mocha', 3000],
            ['yuza_tea', 2500],
            ['green_tea', 2500],
            ['choco_latte', 3000]
        ]

    # 메뉴 출력 메서드
    def menu_print(self):

        for i, item in enumerate(self.menu):
            print(i + 1, item[0], ' : ', item[1], '원')

    #메뉴주문 메서드
    def menu_select(self):

        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

        n = 0
        while len(self.menu)< n or n < 1: # 메뉴에 없는 숫자를 입력하기 전까지 반복한다.
            n = int(input('메뉴 번호를 입력하세요. : '))

            if 1 <= n <= len(self.menu): # 메뉴에 있는 번호를 입력했을 때
                self.order_price.append(self.menu[n-1][1])  # 가격 리스트에 추가
                self.price_sum = self.menu[n-1][1]  # 합계 금액


            else:
                print('메뉴에 없는 번호입니다. 다시 입력해주세요. ')

        # 온도 체크
        t = 0
        while t != 1 and t != 2: # 핫=1, 아이스=2로 설정할 예정, 1 또는 2가 아닐 때까지 반복한다.
            t = int(input('HOT 음료는 숫자 1을, ICE 음료는 숫자 2를 입력해주세요. : '))
            if t == 1:
                self.temp = 'HOT'
                print(self.temp)
            elif t == 2:
                self.temp = 'ICE'
                print(self.temp)
            else:
                print('1 또는 2를 입력해주세요.')

        self.order_menu.append(self.temp + ' ' + self.menu[n-1][0]) #주문 리스트에 메뉴 추가하기
        print('주문한 음료', n, self.temp,' ', self.menu[n-1][0], ':', self.menu[n-1][1], '원')

        #추가 주문 메서드
        while n != 0: # 0은 지불하기로, 외엔 더 추가주문 하기 <<로 설정할것이니 지불하기 전까지 반복하겠다.
            print()
            n = int(input('추가 주문을 하시려면 음료번호를, 계산하기는 숫자 0를 입력해주세요. : '))
            if 0 < n < len(self.menu) + 1:

                # 추가 주문한 음료도 온도 체크해주기
                t = 0
                while t != 1 and t != 2: # 핫=1, 아이스=2로 설정할 예정, 1 또는 2가 아닐 때까지 반복한다.
                    t = int(input('HOT 음료는 숫자 1을, ICE 음료는 숫자 2를 입력해주세요. : '))
                    if t == 1:
                        self.temp = 'HOT' #추가주문용이므로 지역변수인 temp 사용
                        print(self.temp)
                    elif t == 2:
                        self.temp = 'ICE'
                        print(self.temp)
                    else:
                        print('1 또는 2를 입력해주세요.')

                self.order_menu.append(self.temp + ' ' + self.menu[n-1][0]) # 주문 리스트에 추가 메뉴 추가하기
                self.order_price.append(self.menu[n-1][1])  # 가격 리스트에 추가
                self.price_sum += self.menu[n-1][1] # 추가 주문건을 포함한 합계 금액

                print('추가 주문한 음료', self.temp, self.menu[n-1][0], ':', self.menu[n-1][1], '원\n', '합계 금액 : ', self.price_sum, '원')

            else:   #지불하기 선택했을 때
                if n ==0:
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력
                else:  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")


    # 지불
    def pay(self):

        pay_for = input("지불 수단을 선택하세요. (현금은 'cash' 또는 숫자 1을, 카드는 'card' 또는 숫자 2를 입력하세요) : ")
        if pay_for in ['cash', '1']:
            print('직원을 호출하겠습니다.')
        elif pay_for in ['card', '2']:
            print('IC칩 방향에 맞게 카드를 꽂아주세요')
        else:
            print("다시 결제를 시도해 주세요.")


    # 주문서 양식 데코레이터로 만들기
    def receipt(func):
        def line(self): # 클래스 안에 데코레이터를 선언할 때 객체 self 잊지말기
            print('⟝' + '-' * 30 + '⟞')
            for i in range(5):
                print('|' + ' ' * 31 + '|')

            func(self)

            for i in range(5):
                print('|' + ' ' * 31+ '|')
            print('⟝' + '-' * 30 + '⟞')
        return line

    # 주문서 내용
    @receipt # 데코레이터 사용해보기
    def table(self):

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', sum(self.order_price))

        # 주문 일시
        order_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print('주문 일시:', order_time)





a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력2
````
