### ìƒì„±ë°ì´í„° ì••ì¶• í’€ê¸°

````python
import os
import zipfile

def unzip_files(base_path):
    # ê° í´ë” ì´ë¦„ (scissor, rock, paper)
    folders = ['scissor', 'rock', 'paper']

    for folder in folders:
        # í´ë” ë‚´ì˜ ëª¨ë“  íŒŒì¼ ì¤‘ .zip íŒŒì¼ì„ ì œì™¸í•˜ê³  ì‚­ì œ
        folder_path = os.path.join(base_path, folder)
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            if not filename.endswith('.zip') and os.path.isfile(file_path):
                os.remove(file_path)
                # print(f"{filename} ì‚­ì œ ì™„ë£Œ")

        # .zip íŒŒì¼ ê²½ë¡œ
        zip_file_path = os.path.join(base_path, folder, folder + '.zip')

        # Zip íŒŒì¼ì´ ì¡´ì¬í•˜ëŠ” ê²½ìš° ì••ì¶• í•´ì œ
        if os.path.exists(zip_file_path):
            with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                zip_ref.extractall(os.path.join(base_path, folder))
            print(folder + " ì••ì¶• í•´ì œ ì™„ë£Œ")
        else:
            print(zip_file_path + " íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")


base_path = '../data/rock_scissor_paper'
unzip_files(base_path)
````

- ì†ê¸€ì”¨ ëª¨ë¸ì—ì„œ ì‚¬ìš©í•œ ë°ì´í„°ëŠ” (28,28) í˜•íƒœì˜ ë°ì´í„°ì˜€ì§€ë§Œ, ì§€ê¸ˆì€ (224,224) í˜•íƒœì˜ ë°ì´í„° ì…ë‹ˆë‹¤.
- zipíŒŒì¼ ì•ˆì— ìˆëŠ” ë°ì´í„°ê°™ì€ shapeì˜ ë°ì´í„°ë¡œ ë³€ê²½í•˜ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.

````python
def resize_images(img_path):
    images = glob.glob(img_path + "/*.jpg")
    print(len(images), "ê°œì˜ ì´ë¯¸ì§€ê°€ ë¦¬ì‚¬ì´ì¦ˆë  ì˜ˆì •ì…ë‹ˆë‹¤.")

    # ê° íŒŒì¼ì„ ì§€ì •ëœ í¬ê¸°ë¡œ ë³€ê²½
    target_size = (28, 28)
    for img in images:
        old_img = Image.open(img)
        new_img = old_img.resize(target_size)  # Image.ANTIALIAS ì œê±°
        new_img.save(img, "JPEG")

    print(len(images), "ê°œì˜ ì´ë¯¸ì§€ê°€ ë¦¬ì‚¬ì´ì¦ˆë˜ì—ˆìŠµë‹ˆë‹¤.")

image_dir_paths = [
    "../data/rock_scissor_paper/scissor",
    "../data/rock_scissor_paper/rock",
    "../data/rock_scissor_paper/paper"
]

for path in image_dir_paths:
    resize_images(path)
    if path.endswith("/scissor"):
        print("-" * 30)  # 30ê°œì˜ '-'ë¡œ ì´ë£¨ì–´ì§„ ê²½ê³„ì„ 
        print("ê°€ìœ„ ì´ë¯¸ì§€ ë¦¬ì‚¬ì´ì¦ˆ ì™„ë£Œ!")
        print("-" * 30)  # ê²½ê³„ì„  ë‹¤ì‹œ ì¶œë ¥
    elif path.endswith("/rock"):
        print("-" * 30)
        print("ë°”ìœ„ ì´ë¯¸ì§€ ë¦¬ì‚¬ì´ì¦ˆ ì™„ë£Œ!")
        print("-" * 30)
    elif path.endswith("/paper"):
        print("-" * 30)
        print("ë³´ ì´ë¯¸ì§€ ë¦¬ì‚¬ì´ì¦ˆ ì™„ë£Œ!")
        print("-" * 30)
````

````python
import numpy as np

def load_data(img_path, number_of_data, dataset_type="Training"):
    img_size = 28
    color = 3
    imgs = np.zeros(number_of_data * img_size * img_size * color, dtype=np.int32).reshape(number_of_data, img_size, img_size, color)
    labels = np.zeros(number_of_data, dtype=np.int32)

    idx = 0
    for file in glob.iglob(img_path+'/scissor/*.jpg'):
        if idx >= number_of_data:
            break
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:] = img
        labels[idx] = 0
        idx += 1

    for file in glob.iglob(img_path+'/rock/*.jpg'):
        if idx >= number_of_data:
            break
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:] = img
        labels[idx] = 1
        idx += 1

    for file in glob.iglob(img_path+'/paper/*.jpg'):
        if idx >= number_of_data:
            break
        img = np.array(Image.open(file), dtype=np.int32)
        imgs[idx,:,:,:] = img
        labels[idx] = 2
        idx += 1

    print(dataset_type + " ë°ì´í„°ì˜ ë¡œë“œëœ ì´ë¯¸ì§€ ìˆ˜: ", idx)
    return imgs, labels

# í•™ìŠµ ë°ì´í„° ì„¸íŠ¸ ë¡œë“œ
image_dir_path = "../data/rock_scissor_paper"
(x_train, y_train) = load_data(image_dir_path, number_of_data = 300, dataset_type="í•™ìŠµ")
x_train = x_train[:300]
x_train_norm = x_train / 255.0 # ì…ë ¥ì€ 0~1 ì‚¬ì´ì˜ ê°’ìœ¼ë¡œ ì •ê·œí™”

print("x_train shape: {}".format(x_train.shape))
print("y_train shape: {}".format(y_train.shape))
````

# ë”¥ëŸ¬ë‹ ë„¤íŠ¸ì›Œí¬ ì„¤ê³„í•˜ê¸°

````python
import tensorflow as tf
from tensorflow import keras
import numpy as np

# ì´ë¯¸ì§€ ì±„ë„ì´ 3ê°œë¡œ ë°”ë€Œì—ˆê³ , ì¶”ë¡ í•  í´ë˜ìŠ¤ë„ 3ê°œë¡œ ë³€í–ˆìŒ

# ëª¨ë¸ ìƒì„±
model = keras.models.Sequential()

# ì²« ë²ˆì§¸ ì»¨ë³¼ë£¨ì…˜ ì¸µ ì¶”ê°€
model.add(keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(28, 28, 3)))

# ì²« ë²ˆì§¸ ë§¥ìŠ¤ í’€ë§ ì¸µ ì¶”ê°€
model.add(keras.layers.MaxPooling2D(2, 2))

# ë‘ ë²ˆì§¸ ì»¨ë³¼ë£¨ì…˜ ì¸µ ì¶”ê°€
model.add(keras.layers.Conv2D(32, (3, 3), activation='relu'))

# ë‘ ë²ˆì§¸ ë§¥ìŠ¤ í’€ë§ ì¸µ ì¶”ê°€
model.add(keras.layers.MaxPooling2D(2, 2))

# ë°ì´í„° í‰íƒ„í™” ì¸µ ì¶”ê°€
model.add(keras.layers.Flatten())

# ì™„ì „ ì—°ê²° (Dense) ì¸µ ì¶”ê°€
model.add(keras.layers.Dense(64, activation='relu'))

# ì¶œë ¥ ì¸µ ì¶”ê°€ (í´ë˜ìŠ¤ê°€ 3ê°œì„ì„ ê³ ë ¤)
model.add(keras.layers.Dense(3, activation='softmax'))

# ëª¨ë¸ ìš”ì•½ ì¶œë ¥
model.summary()
````

## ë”¥ëŸ¬ë‹ ë„¤íŠ¸ì›Œí¬ í•™ìŠµì‹œí‚¤ê¸°

````python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# ëª¨ë¸ í•™ìŠµ
history = model.fit(x_train, y_train, epochs=10, validation_split=0.2)
````

````python
import numpy as np
import random
import matplotlib.pyplot as plt

# í´ë˜ìŠ¤ë³„ë¡œ ë°ì´í„° ì¸ë±ìŠ¤ë¥¼ ì¶”ì¶œí•˜ëŠ” í•¨ìˆ˜
def extract_class_samples(x, y, num_classes):
    class_indices = {}
    for class_idx in range(num_classes):
        class_indices[class_idx] = [i for i, label in enumerate(y) if label == class_idx]
    return class_indices

# ê° í´ë˜ìŠ¤ë³„ í•˜ë‚˜ì˜ ìƒ˜í”Œì„ ë¬´ì‘ìœ„ë¡œ ì„ íƒí•˜ëŠ” í•¨ìˆ˜
def choose_random_samples(class_indices, num_classes):
    sample_indices = []
    for class_idx in range(num_classes):
        sample_indices.append(random.choice(class_indices[class_idx]))
    return sample_indices

# í•™ìŠµ ë° í…ŒìŠ¤íŠ¸ ë°ì´í„°ì—ì„œ í´ë˜ìŠ¤ë³„ë¡œ ìƒ˜í”Œ ì„ íƒ
num_classes = len(np.unique(y_train)) # í´ë˜ìŠ¤ì˜ ê°œìˆ˜ë¥¼ ê²°ì •
train_class_indices = extract_class_samples(x_train, y_train, num_classes)
test_class_indices = extract_class_samples(x_test, y_test, num_classes)

# ë¬´ì‘ìœ„ë¡œ í•™ìŠµ ë° í…ŒìŠ¤íŠ¸ ìƒ˜í”Œ ì„ íƒ
train_samples = choose_random_samples(train_class_indices, num_classes)
test_samples = choose_random_samples(test_class_indices, num_classes)

# ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ ì˜¬ë°”ë¥¸ í˜•ì‹ìœ¼ë¡œ ë¡œë“œ ë° ë³€í™˜
x_train = np.array(x_train)  # ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ numpy ë°°ì—´ë¡œ ë³€í™˜
x_test = np.array(x_test)    # ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ numpy ë°°ì—´ë¡œ ë³€í™˜

# ì‹œê°í™”
fig, axes = plt.subplots(2, num_classes, figsize=(15, 4))

for i, idx in enumerate(train_samples):
    axes[0, i].imshow(x_train[idx], cmap=plt.cm.binary)
    axes[0, i].set_title(f'Class {y_train[idx]}')
    axes[0, i].axis('off')

for i, idx in enumerate(test_samples):
    axes[1, i].imshow(x_test[idx], cmap=plt.cm.binary)
    axes[1, i].set_title(f'Class {y_test[idx]}')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
````
