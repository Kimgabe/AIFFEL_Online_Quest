# ✨ 메인 퀘스트 1번 [계좌 클래스 만들기]

```python
class Account :
    # 메인 퀘스트 1번 Q2(계좌 객체 개수 저장)
    count = 0

    # 메인 퀘스트 1번 Q1
    def __init__(self) :
        self.account_init = True
        self.name = input("예금주를 입력해주세요 >> ")
        self.bank = "SC은행"

        # 숫자 외 입력시 처리
        while self.account_init :
            try :
                self.money = int(input("초기 입금액을 입력해주세요. >> "))
            except :
                print("숫자를 입력해주세요.")
                continue
            else :
                self.account_init = False


        # 10-100-100000 -> 10-100-100001 순으로 생성(Account.count // 810000000이 90이 되는 순간 자리수 바뀌지만 어마어마한 숫자라 현재는 무시)
        self.account_idx = f"{10 + (Account.count // 810000000)}-{100 + (Account.count // 900000) % 900}-{100000 + (Account.count % 900000)}"

        # 메인 퀘스트 1번 Q2(계좌 객체 개수 저장)
        Account.count += 1

        # 메인 퀘스트 1번 Q7(5회 입금 시 이자 지급)
        ##### self.deposit_count = 0
        # 메인 퀘스트 1번 Q10(입출금 내역)
        self.deposit_history = []
        self.withdraw_history = []

    # 메인 퀘스트 1번 Q3
    def get_account_num(self) :
        return Account.count

    # 메인 퀘스트 1번 Q4(입금)
    def deposit(self) :
        self.deposit_init = True
        while self.deposit_init :
            try :
                self.deposit = int(input("입금할 금액을 입력해주세요. >> "))
            except :
                print("숫자를 입력해주세요.")
            else :
                if self.deposit < 1 :
                    print("1원 이상만 입금 가능합니다.")
                else :
                    self.money += self.deposit
                    self.deposit_history.append(self.deposit) # 메인 퀘스트 1번 Q10(입출금 내역)
                    self.deposit_init = False
                    ##### self.deposit_count += 1
                    print(f"{format(self.deposit, ',')}원 입금되었습니다.")
                    del self.deposit # 반복 시 callable 오류 방지

                    # 메인 퀘스트 1번 Q7(5회 입금 시 이자 지급)
                    ##### if self.deposit_count % 5 == 0 :
                    if len(self.deposit_history) % 5 == 0 :
                        self.interest = round(0.01*self.money)
                        self.money += self.interest
                        print(f"입금 횟수가 5회가 되어 이자 {format(self.interest, ',')}원 입금되었습니다.")

                    print("="*20)
                    print(f"현재 잔고는 {format(self.money, ',')}원 입니다.")
                    print("="*20)

    # 메인 퀘스트 1번 Q5(출금)
    def withdraw(self) :
        self.withdraw_init = True
        while self.withdraw_init :
            try :
                self.withdraw = int(input("출금할 금액을 입력해주세요. >> "))
            except :
                print("숫자를 입력해주세요.")
            else :
                if self.withdraw < 1 :
                    print("1원 이상을 입력해주세요")
                elif self.withdraw > self.money :
                    print("출금은 계좌의 잔고 이상으로 출금할 수 없습니다.")
                else :
                    self.money -= self.withdraw
                    self.withdraw_history.append(self.withdraw) # 메인 퀘스트 1번 Q10(입출금 내역)
                    self.withdraw_init = False
                    print(f"{format(self.withdraw, ',')}원 출금되었습니다.")
                    del self.withdraw # 반복 시 callable 오류 방지
                    print("="*20)
                    print(f"현재 잔고는 {format(self.money,',')}원 입니다.")
                    print("="*20)

    # 메인 퀘스트 1번 Q6(정보출력)
    def display_info(self) :
        print(f"(은행이름 : {self.bank}, 예금주 : {self.name}, 계좌번호 : {self.account_idx}, 잔고 : {format(self.money,',')}원)")
```

```python
a = Account()

# 메인 퀘스트 1번 Q4(입금)
# 메인 퀘스트 1번 Q7(5회 입금 시 이자 지급)
print()
print("="*5 + "입금 및 이자 지급" + "="*5)
for i in range(5) :
    print()
    a.deposit()

# 메인 퀘스트 1번 Q5(출금)
print()
print("="*5 + "출금" + "="*5)
a.withdraw()

# 메인 퀘스트 1번 Q6(정보출력)
print()
print("="*5 + "정보출력" + "="*5)
a.display_info()
```

# ✨ 메인 퀘스트 2번 [명단 관리]

```python
# 메인 퀘스트 2번
class Aiffel:
    # 메인 퀘스트 2번 Q1(학번 부여)
    year = 2023
    generation = "29기"
    curriculum = "코어"

    def __init__(self, dic:dict) :
        self.people_dic = dic
        self.idx = 0
        for k in self.people_dic.keys() :
            self.idx += 1
            # 사람 이름 키값으로한 딕셔너리 value로 [학번 0, 과정종류 1, 퀘스트점수총점 2, 퀘스트입력총점 3, 퀘스트penalty총합 4] 리스트 생성
            self.people_dic[k] = [f"{str(Aiffel.year)[-2:]}{Aiffel.generation[0:2]}{2 if Aiffel.curriculum == '코어' else 1}{format(self.idx,'02')}",\
                                  Aiffel.curriculum,0,0,0]


    # 메인 퀘스트 2번 Q2(학번 개수 출력)
    def get_id_num(self) :
        # return self.idx  # 학번 부여 시 사용된 인덱스로도 도출 가능
        return len(self.people_dic.values())

    # 메인 퀘스트 2번 Q3(퀘스트 점수 입력)
    def score(self) :
        # 점수부여할 사람
        print()
        self.get_name = input("점수를 주고자 하는 분의 성함을 입력해주세요 >> ")
        if self.get_name not in self.people_dic :
            print("해당하는 사람은 없습니다.")
            return

        # 1~3점 점수 부여
        self.score_init = True
        while self.score_init :
            print()
            print(f"{self.get_name}님에게 몇점을 주시겠습니까?")
            print("1) 1점   2) 2점  3) 3점")

            try :
                print()
                self.get_score = int(input("점수를 선택해주세요. >> "))
            except :
                print("숫자만 입력해주시기 바랍니다.")

            else :
                if 1 <= self.get_score <= 3 :
                    self.people_dic[self.get_name][2] += self.get_score # 퀘스트점수총점
                    self.people_dic[self.get_name][3] += self.get_score # 퀘스트입력총점
                    print()
                    print(f"현재 {self.get_name}님의 총점은 {self.people_dic[self.get_name][2]} 입니다.")
                    del self.get_score
                    self.score_init = False
                else :
                    print("해당하는 숫자만 입력해주시기 바랍니다.")

        del self.get_name

    # 메인 퀘스트 2번 Q4(퀘스트 점수 차감)
    def penalty(self) :
        # 이름
        print()
        self.get_name = input("점수를 차감하고자 하는 분의 성함을 입력해주세요 >> ")

        if self.get_name not in self.people_dic :
            print("해당하는 사람은 없습니다.")
            return

        # 점수 차감
        self.penalty_init = True
        while self.penalty_init :
            print()
            print(f"{self.get_name}님에게 몇점을 차감하시겠습니까?")

            try :
                print()
                self.get_penalty = int(input("점수를 입력해주세요. >> "))
            except :
                print("숫자만 입력해주시기 바랍니다.")

            else :
                self.people_dic[self.get_name][2] -= abs(self.get_penalty) # 퀘스트점수총점
                self.people_dic[self.get_name][4] += abs(self.get_penalty) # 퀘스트penalty총합
                print()
                print(f"현재 {self.get_name}님의 총점은 {self.people_dic[self.get_name][2]} 입니다.")
                del self.get_name
                self.penalty_init = False

        del self.get_penalty

    # 메인 퀘스트 2번 Q5(CSV파일 출력)
    def display_info(self) :
        import pandas as pd
        df = pd.DataFrame(self.people_dic).transpose()
        df = df.reset_index()
        df.columns = ["이름", "학번", "과정종류", "퀘스트점수총점", "퀘스트입력총점", "퀘스트penalty총합"]
        df.to_csv("새로운 코어 명단.csv", mode='w', header=True)
        print("새로운 코어 명단.csv으로 저장하였습니다.")

    # 메인 퀘스트 2번 Q6(중도 포기자 삭제)
    def delete_people(self) :
        print()
        self.get_name = input("삭제하고자 하는 분의 성함을 입력해주세요 >> ")

        if self.get_name in self.people_dic.keys() :
            print(f"{self.get_name}님을 명단에서 삭제하겠습니다.")
            del self.people_dic[self.get_name]
            del self.get_name
        else :
            print("올바른 성명을 입력해주시기 바랍니다.")
            print(", ".join(self.people_dic.keys()))
```

```python
# 메인 퀘스트 2번 Q1(학번 부여)
people_dic = {"강영현":0,"김경민":0,"김민기":0,"김승순":0,"김양희":0,\
              "김영신":0,"오우진":0,"윤혜윤":0,"이상원":0,"이선재":0,\
                "이슬":0,"이승환":0,"전민규":0,"조보겸":0,"조수민":0,"최강훈":0,"한현종":0,"홍채림":0}

c = Aiffel(people_dic)

# 메인 퀘스트 2번 Q2(학번 개수 출력)
print(a.get_id_num())

# 메인 퀘스트 2번 Q3(점수 입력)
c.score()

# 메인 퀘스트 2번 Q4(점수 차감)
c.penalty()

# 메인 퀘스트 2번 Q6(중도 포기자 삭제)
c.delete_people()
c.delete_people()

# 메인 퀘스트 2번 Q5(CSV파일 출력)
a.display_info()
```

```python
# 메인 퀘스트 2번 Q7, Q10(그룹 클래스)
class Aiffel_Group(Aiffel) :

  def __init__(self, dic) :
    super().__init__(dic)
    self.people_list = list(self.people_dic.keys())
    self.groups = ["상생","신뢰","열정","이끔","성장","가치","공유","확산"]

  def grouping(self) :
    import random
    numbers = list(range(len(self.people_list)))
    random.shuffle(numbers)

    self.idx = 0
    self.groups_people = {}
    for i in numbers :
      if self.groups[self.idx] in self.groups_people.keys() :
        self.groups_people[self.groups[self.idx]].append(self.people_list[i])
      else :
        self.groups_people[self.groups[self.idx]] = [self.people_list[i]]

      # 인덱스 초기화
      if self.idx < len(self.groups)-1 :
        self.idx += 1
      else :
        self.idx = 0

    return self.groups_people

    # 길드 딕셔너리 받기
    def group2guild_score(self, name, score) :
      self.select_group = name
      self.select_score = score
```

```python
q7 = Aiffel_Group(people_dic)

# 메인 퀘스트 2번 Q7(그룹 클래스)
print(q7.grouping())
```

```python
# 메인 퀘스트 2번 Q8(그룹 클래스)
class Aiffel_Guild(Aiffel) :
  def __init__(self, dic) :
    super().__init__(dic)
    self.people_list = list(self.people_dic.keys())
    self.guilds = ["힌튼", "응", "하사비스"]
    self.guilds_score = [0,0,0]

  def guilding(self) :
    import random
    numbers = list(range(len(self.people_list)))
    random.shuffle(numbers)

    self.idx = 0
    self.guilds_people = {}
    for i in numbers :
      if self.guilds[self.idx] in self.guilds_people.keys() :
        self.guilds_people[self.guilds[self.idx]].append(self.people_list[i])
      else :
        self.guilds_people[self.guilds[self.idx]] = [self.people_list[i]]

      # 인덱스 초기화
      if self.idx < len(self.guilds)-1 :
        self.idx += 1
      else :
        self.idx = 0

    return self.guilds_people

  def guild_score(self) :

    if len(self.guilds_people) == 0 :
      print("그룹핑을 먼저 하십시오.")
      return


    self.guild_score_init = True
    while self.guild_score_init:
      print()
      print("다음 중 어느 길드에 점수를 변경하시겠습니까?")
      print(" ".join([f"{i+1}. {j}" for i, j in enumerate(self.guilds)]))

      try :
        print()
        self.answer = int(input("번호를 선택해주세요 >> "))
      except :
        print("숫자를 입력해주세요.")
      else :
        if 1 <= self.answer <= 3 :
          self.guild_score_answer_init = True
          while self.guild_score_answer_init :
            try :
              print()
              self.answer_score = int(input("증가 또는 차감하실 점수를 입력해주세요(-시 차감) >> "))
            except :
              print("숫자를 입력해주세요.")
            else :
              if self.answer_score < 0 :
                print(f"{self.guilds[self.answer-1]}그룹의 점수를 {abs(self.answer_score)}점을 차감하겠습니다.")
              elif self.answer_score > 0 :
                print(f"{self.guilds[self.answer-1]}그룹의 점수를 {abs(self.answer_score)}점을 증가하겠습니다.")

              # 그룹 총점 변경
              self.guilds_score[self.answer-1] += self.answer_score
              print(f"{self.guilds[self.answer-1]}그룹의 현재 총점은 {self.guilds_score[self.answer-1]}입니다.")

              # 그룹 인원들 전체 총점 변경
              for i in self.guilds_people[self.guilds[self.answer-1]] :
                # 멤버이름(퀘스트 점수 총점 조정)
                people_dic[i][2] += self.answer_score

              del self.answer
              del self.answer_score
              self.guild_score_init = False
              self.guild_score_answer_init = False
```

```python
q8 = Aiffel_Guild(people_dic)

# 메인 퀘스트 2번 Q8(그룹 클래스)
print(q8.guilding())

# 메인 퀘스트 2번 Q9(점수 부여)
q8.guild_score()
```
