```python
# open("https://github.com/NeatyNut/AIFFEL_Online_Quest/blob/main/Quest03/Avengers.txt")
f = open('Avengers.txt')
text = f.read()
f.close()
text
```

```python
import re
from itertools import chain

def list_split(s, word_list) :
  for index, i in enumerate(word_list) :
    if s in i :
      word_list[index] = i.split(s)
    else :
      word_list[index] = [i]
  return word_list

text = text.lower()
text = re.sub(r'[\n\r]', ' ', text)
word_list = text.split(" ")
list_split("'", word_list)
word_list = list(chain(*word_list))
word_list = list(map(lambda x : re.sub(r'[^\w\s]', '', x), word_list))
word_list = list(filter(lambda x: x != '', word_list))
word_list
```

```python
count_list = []
for i in range(0, len(word_list), 2) :
  if i != len(word_list)-1 :
    count_list.append((word_list[i], word_list[i+1]))

count_list
```

```python
lis = set(count_list)
answer = {}
for i in lis :
  answer[i] = count_list.count(i)

max_key = max(answer, key=answer.get)
max_value = answer[max_key]

print(max_key)
print(max_value)
```
